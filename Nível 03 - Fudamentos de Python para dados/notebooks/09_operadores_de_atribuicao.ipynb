# **Operadores de atribuição**

**Operadores de Atribuição:** Os operadores de atribuição são usados para armazenar valores em variáveis, e também para atualizar esses valores com base em operações matemáticas ou lógicas.

```python
# Operador de atribuição simples (=)
nome = 'João'
idade = 25
```

```python
# Operador de atribuição com soma (+=)
qtd = 5

# qtd = qtd + 3
# print(qtd)

qtd += 3
print(qtd)
```

```python
# Operador de atribuição com subtração (-=)
qtd = 8

# qtd = qtd - 2
# print(qtd)

qtd -= 2
print(qtd)
```

```python
# Operador de atribuição com multiplicação (*=)
qtd = 2

# qtd = qtd * 4
# print(qtd)

qtd *= 4
print(qtd)
```

```python
# Operador de atribuição com divisão (/=)
qtd = 10

# qtd = qtd / 5
# print(qtd)

qtd /= 5
print(qtd)
```

```python
# Operador de atribuição com divisão exata (//=)
qtd = 7

# qtd = qtd // 2
# print(qtd)

qtd //= 2
print(qtd)
```

```python
# Operador de atribuição com divisão potência (**=)
qtd = 4

# qtd = qtd ** 3
# print(qtd)

qtd **= 3
print(qtd)
```

```python
# Operador de atribuição com divisão resto da divisão (%=)
qtd = 8

# qtd = qtd % 3
# print(qtd)

qtd %= 3
print(qtd)
```
