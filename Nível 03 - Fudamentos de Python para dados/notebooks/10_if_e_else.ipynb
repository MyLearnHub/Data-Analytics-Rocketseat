# **If e Else**

**Estruturas Condicionais:** As estruturas condicionais são utilizadas para que o programa tome uma decisão com base em uma condição, que pode ser verdadeira ou falsa. Caso a condição seja verdadeira, ele realiza alguma instrução, caso seja falsa, ele realiza outra instrução.

```python
# Sintaxe do comando if
'''
if condição: (Se a condição for verdadeira)
    Executa esse bloco de comandos
'''
```

```python
# Exemplo com if
idade = 15

if idade >= 18:
    print('Você é maior de idade!')

print('Indentação')
```

```python
# Sintaxe do comando if-else
'''
if condição: (Se a condição for verdadeira)
    Executa esse bloco de códigos
else: (Se a condição do if for falsa)
    Executa esse bloco de códigos
'''
```

```python
# Exemplo com if-else
idade = 13

if idade >= 18:
    print('Você é maior de idade!')
else:
    print('Você é menor de idade!')
```
