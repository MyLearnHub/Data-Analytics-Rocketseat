# **While True, Break e Continue**

```python
# Instrução while True
'''
while True:
    Executa o bloco de códigos
    if condição: (Se a condição for verdadeira)
        Executa esse bloco (break)
'''
```

```python
while True:
    senha = input('Digite a sua senha: ')
    if senha == 'python':
        print('Acesso liberado!')
        break
    else:
        print('Senha incorreta! Digite novamente!')
```

```python
# Instrução while True - Leia números e imprima na tela até o usuário digitar o número 0.
while True:
    numeros = int(input('Digite um número inteiro: '))
    if numeros == 0:
        print('Programa encerrado!')
        break
    else:
        print('O número digitado foi: ', numeros)
```

```python
# Comando break
i = 1

while i < 10:
    print(i)
    if i == 3:
        break
    i += 1
```

```python
# Comando continue - Imprima de 1 a 20, com exceção do número 10
x = 1

while x < 21:
    x += 1
    if x == 10:
        continue
    else:
        print(x)
```
