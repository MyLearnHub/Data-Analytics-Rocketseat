# **Formatação de Strings**

**F-Strings:** As f-strings são uma maneira de inserir variáveis e expressões diretamente dentro de uma string, de forma fácil e legível.

```python
nome = input('Digite o seu nome: ')
idade int(input('Digite a sua idade: '))

print('O seu nome é: ', nome, 'e a sua idade é: ', idade)
```

```python
# Formatação de strings
nome = input('Digite o seu nome: ')
idade int(input('Digite a sua idade: '))

print(f'O seu nome é {nome} e a sua idade é {idade}')
```

```python
# Formatar números float
preco = 49.95823

print(f'O preço do produto é R${preco:.2f}')
```

```python
# Formatar números inteiros (preenchimento com zeros)
numero = 7
print(f'O número é: {numero:07}')
```

```python
# Alinhar texto: à esquerda, à direita e centralizado
texto = 'Python'
print(f'{texto:<10} - alinhado à esquerda')
print(f'{texto:>10} - alinhado à direita')
print(f'{texto:^10} - alinhado ao centro')
```

```python
# Inserir separadores de milhar
valor = 1234567
print(f'Valor formatado: {valor:,}')
```

```python
# Exibir porcentagens
taxa = 0.085
print(f'Taxa de crescimento: {taxa:.2%}')
```

```python
# Representar números em diferentes bases (binário, hexadecimal, octal)
numero = 255

print(f'Binário: {numero:b}')
print(f'Octal: {numero:o}')
print(f'Hexadecimal: {numero:x}')
```

```python
# Cálculos e funções diretamente nas chaves
a = 5
b = 3

print(f'A soma entre os números {a} + {b} é = {a + b}')
```

```python
# Usando métodos de string
nome = 'Python'
print(f'Nome em maíusculo: {nome.upper()}')
```

```python
# Apresentando o faturamento de uma empresa
nome_da_empresa = 'DataCorp'
faturamento = 1578934.58

print(f'A empresa {nome_da_empresa} faturou R$ {faturamento:,.1f} neste mês')
```
