# **Operadores relacionais**

**Operadores Relacionais:** Os operadores relacionais servem para comparar valores. O resultado de uma comparação sempre será um valor booleano: True (verdadeiro) ou False (falso).

```python
# Operador de igualdade
a = 10
b = 10

a == b
```

```python
# Operador de igualdade com strings
nome = 'José'

nome == 'José' # Case Sensitive
```

```python
# Maior que
a = 5
b = 5

a > b
b > a
```

```python
# Maior ou igual a
a = 7
b = 7

a >= b
b >= a
```

```python
# Menor que
a = 7
b = 5

a < b
b < a

a < 10
```

```python
# Menor ou igual a
a = 14
b = 14

a <= b
b <= a
```

```python
# Diferente de
a = 10
b = 10

a != b
```

```python
# Diferente de com strings
fruta = 'Laranja'

fruta != 'Morango'
```
