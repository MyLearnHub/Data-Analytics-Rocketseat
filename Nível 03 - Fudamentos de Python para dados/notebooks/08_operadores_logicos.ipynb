# **Operadores lógicos**

**Operadores Lógicos:** Os operadores lógicos são usados para conectar duas ou mais expressões booleanas, ou seja, expressões que retornam True ou False.

O resultado final também será um valor booleano.

```python
# Operador E
idade = 19
tem_carteira = False

idade >= 18 and tem_carteira

'''
True and True = True
True and False = False
False and True = False
False and False = False
'''
```
