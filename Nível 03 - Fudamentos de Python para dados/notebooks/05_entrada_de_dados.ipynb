# **Entrada de dados em Python**

```python
# Comando input para entrada de dados em Python
nome = input('Digite o seu nome: ')
print(nome)
```

```python
# Verificando o tipo da variável nome
type(nome)
```

```python
# Lendo um número inteiro
idade = input('Digite a sua idade: ')
print(idade)
```

```python
# Verificando o tipo da variável idade
type(idade)
```

```python
# Lendo e convertendo para inteiro
idade = int(input('Digite a sua idade: '))
print(idade)
type(idade)
```

```python
# Lendo e convertendo para float
altura = float(input('Digite a sua altura: '))
print(altura)
type(altura)
```

```python
# Exemplo utilizando input e print
produto = input('Digite o nome do produto: ')
preco = float(input('Digite o preço do produto: '))
quantidade = int(input('Digite a quantidade do produto em estoque: '))

print('Nome do produto: ', produto, ' Preço: ', preco, ' Quantidade em estoque: ', quantidade)
```
