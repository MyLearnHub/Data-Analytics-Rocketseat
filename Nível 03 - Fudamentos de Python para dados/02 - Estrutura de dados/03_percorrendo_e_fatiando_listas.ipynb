# **Percorrendo e fatiando listas**

```python
# Percorrendo uma lista com for
carros = ["gol", "saveiro", "compass"]
print(carros)

for x in carros:
  print(x)
```

```python
# Percorrendo usando range de listas
for x in range(len(carros)):
  print(x)
```

```python
# Utilizar o while
contador = 0
while contador < len(carros):
  print(carros[contador])
  contador += 1
```

```python
# Fatiamento de listas
compras = ["arroz", "feijão", "farinha de trigo", "açúcar", "manteiga"]
print(compras)

print(compras[0:2])
print(compras[::])
print(compras[-1])
print(compras[2:4])
print(compras[1:6:2])
print(compras[-1::-1])
```
