# **Retornando chaves e valores nos dicionários**

```python
# Acessando valores de um dicionário (indíce, get)
pessoas = {
    "nome": "Sonia dos Sandos",
    "idade": 32,
    "altura": 1.69,
    "habilitado": False
}

# print(pessoas)

# print(pessoas("habilitado"))

print(pessoas.get("idade"))
```

```python
# Retornando as chaves de um dicionário
print(pessoas.keys())
```

```python
# Retornando os valores de um dicionário
print(pessoas.values())
```

```python
# Retornando chave e valor
print(pessoas.items())
```

```python
# Verificando a existência de ua chave específica
"cpf" in pessoas
```

```python
# Alterando valores no dicionário
pessoas["idade"] = 40
print(pessoas)

pessoas.update({"altura":1.88})
print(pessoas)

pessoas = {
    "nome": "Sonia dos Sandos",
    "idade": 32,
    "altura": 1.69,
    "habilitado": False
}

# pessoas["idade"] = 42
# pessoas["altura"] = 1.88
# pessoas["habilitado"] = True

pessoas.update({"idade": 42, "altura": 1.88, "habilitado": True})
print(pesosas)
```

```python
# Adicionando um novo chave:valor no dicionário
pessoas["cpf"] = "000.000.000-00"
print(pessoas)

pesosas.update({"rg":"55.555.555-55"})
print(pessoas)
```

```python
# Removendo um item chave:valor no dicionário
pessoas.pop("rg")
print(pessoas)
```

```python
# Removendo o último elemento inserido
pessoas.popitem()
print(pessoas)
```

```python
# Utilizando o del
cliente = {
    "nome": "Priscila Costa",
    "data_nasc": "28-04-1996",
    "cpf": "000.000.000-00".
    "rg": "55.555.555-55"
}

print(cliente)

del cliente["rg"]
print(cliente)
```

```python
del cliente
print(cliente)
```

```python
cliente.clear()
print(cliente)
```
