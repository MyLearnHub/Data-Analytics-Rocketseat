# **Métodos de listas**

```python
# Tamanho de uma lista
cores = ["azul", "amarelo", "vermelho"]
print(cores)

print(len(cores))
```

```python
# Inserindo dados em uma lista (append)
cores.append("lilás")
print(cores)

cores.append("marrom")
print(cores)
```

```python
# Inserindo dados em uma posição específica (insert)
frutas = ["laranja", "goiaba", "abacate"]
print(frutas)

frutas.insert(0, "maracujá")
print(frutas)

frutas.insert(2, "jabucticaba")
print(frutas)
```

```python
# Juntando listas com o extend
cores1 = ["azul", "amarelo"]
cores2 = ["vermelho", "leranja"]

print(cores1)
print(cores2)

cores1.extend(cores2)

print(cores1)
```

```python
# Removendo dados de uma lista (remove)
frutas = ["laranja", "goiaba", "abacate"]
print(frutas)

# frutas.remove("goiaba")

print(frutas)
```

```python
# Removendo dados de uma lista (pop)
# frutas.pop()

print(frutas)

frutas.pop(0)

print(frutas)
```

```python
# Removendo de uma lista (del)
frutas = ["laranja", "goiaba", "abacate"]
print(frutas)

del frutas[0]
print(frutas)
```

```python
# Excluindo uma lista(del)
del frutas
print(frutas)
```

```python
# Limpando todos os items de uam lista (clear)
frutas = ["laranja", "goiaba", "abacate"]

print(frutas)

frutas.clear()
print(frutas)
```
