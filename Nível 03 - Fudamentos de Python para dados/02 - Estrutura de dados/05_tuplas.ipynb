# **Tuplas**

**TUPLAS:** Uma tupla é uma estrutura de dados utilizada também para armazenar mais de um dado por vez na mesma estrutura. Porém, ela tem caracteristicas diferentes de uma lista.

* Tuplas são imutáveis, ou seja, não é possível alterar um elemento que faz parte dela.

```python
# Definindo uma tupla (SINTAXE)
carros = ("Gol", "Voyage", "Fusca")
print(carros)
```

```python
# Verificando o tipo de dados
print(type(carros))
```

```python
# Acessando um elemento de uma tupla
print(carros[0])
print(carros[1])
print(carros[2])
print(carros[-1])
```

```python
# Tuplas são imutáveis
carros[0] = "hilux"
```

```python
# Alterando dados de uma tupla
frutas = ("abacaxi", "banana", "limão", "laranja", "maracujá", "maçã")
print(frutas)

auxiliar = list(frutas)

print(type(auxiliar))
print(auxiliar)

auxiliar[1] = "kiwi"
print(auxiliar)

frutas = tupla(auxiliar)
print(frutas)
```

```python
# Posso ter elementos iguais dentro de uma tupla
frutas = ("abacaxi", "maçã", "limão", "abacaxi")
print(frutas)
```

```python
# Verificando o tamanho de uma tupla
print(len(frutas))
```

```python
# Extraindo dados de uma tupla em variáveis
numeros = (10, 50, 14)

(n1, n2, n3) = numeros

print(f"1º valor {n1}, 2º {n2}, 3º {n3}")

numeros = (10, 50, 14, 9, 237, 80, 41)

(n1, n2, *n3) = numeros
print(n1)
print(n2)
print(n3)
```

```python
# Percorrendo uma tupla
frutas = ("abacaxi", "maçã", "limão", "goiaba")

for x in frutas:
  print(x)
```

```python
# Juntando tuplas
frutas1 = ("morango", "kiwi")
frutas2 = ("maçã", "banana")

frutas3 = frutas1 + frutas2

print(frutas1)
print(frutas2)
print(frutas3)
```
