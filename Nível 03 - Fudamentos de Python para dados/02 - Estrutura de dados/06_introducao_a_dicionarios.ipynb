# **Introdução a dicionários**

Um dicionário em Python é uma coleção ordenada (A partir da versão 3.7) é mutável, composta por pares de chaves e valor (key: value).

```python
# Sintaxe básica
carros = {}

pessoas = {
    "nome": "Joaquim da Silva",
    "idade": 32,
    "altura": 1.88,
    "habilitado": True
}

print(pessoas)
print(type(pessoas))
```

```python
# Acessando um item de um dicionário
print(pessoas["nome"])
print(pessoas["idade"])
print(pessoas["altura"])
print(pessoas["habilitado"])
```

```python
# Não aceita valores duplicados
carro = {
    "nome": "Gol",
    "marca": "Volkswagen",
    "modelo": "G5",
    "modelo": "Trend",
}
print(carro)
```

```python
# Modificando um item de um dicionário
carro = {
    "nome": "Gol",
    "marca": "Volkswagen",
    "modelo": "G5",
}
print(carro)

carro["modelo"] = "G2"
print(carro)
```

```python
# Tamanho de um dicionário
print(len(carro))
```
