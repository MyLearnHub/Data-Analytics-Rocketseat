### Solu√ß√£o 1.1 - Criando e Acessando Listas

**Explica√ß√£o**:
Este exerc√≠cio testa a cria√ß√£o de listas e o acesso a elementos usando √≠ndices positivos e negativos. Tamb√©m verifica a modifica√ß√£o de elementos e o uso do operador `in` para verificar exist√™ncia.

**Conceitos aplicados**:
- Cria√ß√£o de listas
- Acesso por √≠ndices positivos (come√ßando em 0)
- Acesso por √≠ndices negativos (-1 √© o √∫ltimo elemento)
- Operador `in` para verifica√ß√£o de exist√™ncia
- Modifica√ß√£o de elementos em listas

```python
# Criando a lista de frutas
frutas = ['ma√ß√£', 'banana', 'laranja', 'uva', 'morango']

# 1. Exibindo o primeiro elemento usando √≠ndice positivo (0)
print(f"Primeiro elemento: {frutas[0]}")

# 2. Exibindo o √∫ltimo elemento usando √≠ndice negativo (-1)
print(f"√öltimo elemento: {frutas[-1]}")

# 3. Exibindo o terceiro elemento usando √≠ndice positivo (2)
print(f"Terceiro elemento: {frutas[2]}")

# 4. Verificando se 'banana' est√° na lista usando o operador 'in'
esta_na_lista = 'banana' in frutas
print(f"banana est√° na lista: {esta_na_lista}")

# 5. Modificando o segundo elemento (√≠ndice 1) para 'abacaxi'
frutas[1] = 'abacaxi'
print(f"Lista ap√≥s modifica√ß√£o: {frutas}")
```

**üí° Observa√ß√µes importantes**:

- √çndices positivos come√ßam em 0 (primeiro elemento)
- √çndices negativos come√ßam em -1 (√∫ltimo elemento)
- O operador `in` retorna `True` se o elemento existe, `False` caso contr√°rio
- Listas s√£o mut√°veis, ent√£o podemos modificar elementos diretamente usando `lista[√≠ndice] = novo_valor`

**üéØ Varia√ß√µes poss√≠veis**:

- Podemos usar `frutas.index('banana')` para encontrar o √≠ndice de 'banana' (mas gera erro se n√£o existir)
- Podemos verificar m√∫ltiplos elementos: `'banana' in frutas and 'ma√ß√£' in frutas`

### Solu√ß√£o 1.2 - Manipulando Listas com M√©todos

**Explica√ß√£o**:
Este exerc√≠cio combina m√∫ltiplos m√©todos de manipula√ß√£o de listas: adicionar elementos (append, insert, extend), remover elementos (remove, pop) e verificar tamanho (len).

**Conceitos aplicados**:
- M√©todo `append()` para adicionar ao final
- M√©todo `insert()` para inserir em posi√ß√£o espec√≠fica
- M√©todo `extend()` para adicionar m√∫ltiplos elementos
- M√©todo `remove()` para remover por valor
- M√©todo `pop()` para remover e retornar elemento
- Fun√ß√£o `len()` para obter tamanho

```python
# Lista inicial de compras
compras = ['arroz', 'feij√£o', 'macarr√£o']

# 1. Adicionando 'a√ß√∫car' ao final usando append()
compras.append('a√ß√∫car')

# 2. Inserindo '√≥leo' na posi√ß√£o 1 usando insert()
# insert(posi√ß√£o, elemento) - insere antes do elemento na posi√ß√£o especificada
compras.insert(1, '√≥leo')

# 3. Criando outra lista e adicionando todos os elementos usando extend()
outros_itens = ['sal', 'pimenta']
compras.extend(outros_itens)  # Adiciona todos os elementos de outros_itens

# 4. Removendo 'macarr√£o' usando remove()
# remove() remove a primeira ocorr√™ncia do valor especificado
compras.remove('macarr√£o')

# 5. Removendo o √∫ltimo elemento usando pop() e armazenando em vari√°vel
# pop() remove e retorna o √∫ltimo elemento (ou elemento no √≠ndice especificado)
item_removido = compras.pop()

# 6. Exibindo o tamanho da lista usando len()
tamanho = len(compras)

# Exibindo resultados
print(f"Lista final: {compras}")
print(f"Item removido com pop: {item_removido}")
print(f"Tamanho da lista: {tamanho}")
```

**üí° Observa√ß√µes importantes**:

- `append()` adiciona um √∫nico elemento ao final
- `insert(posi√ß√£o, elemento)` desloca elementos existentes para a direita
- `extend()` adiciona todos os elementos de um iter√°vel (lista, tupla, etc.)
- `remove(valor)` gera erro se o valor n√£o existir na lista
- `pop()` sem argumento remove o √∫ltimo elemento; `pop(√≠ndice)` remove elemento no √≠ndice
- `len()` retorna o n√∫mero de elementos na lista

**üéØ Varia√ß√µes poss√≠veis**:

- Podemos usar `compras += outros_itens` em vez de `extend()` (comportamento similar)
- Podemos usar `compras.pop(0)` para remover o primeiro elemento
- Podemos usar `del compras[√≠ndice]` para remover por √≠ndice sem retornar valor

### Solu√ß√£o 1.3 - Percorrendo e Fatiando Listas

**Explica√ß√£o**:
Este exerc√≠cio demonstra diferentes formas de iterar sobre listas e usar fatiamento (slicing) para acessar subconjuntos de elementos.

**Conceitos aplicados**:
- Itera√ß√£o com `for item in lista`
- Itera√ß√£o com `for x in range(len(lista))` para acesso por √≠ndices
- Fatiamento b√°sico: `lista[in√≠cio:fim]`
- Fatiamento com passo: `lista[in√≠cio:fim:passo]`
- Fatiamento reverso: `lista[-1::-1]`

```python
# Lista de notas
notas = [7.5, 8.0, 6.5, 9.0, 7.0, 8.5, 6.0, 9.5]

# 1. Percorrendo a lista usando 'for item in lista'
print("Percorrendo com 'for item in lista':")
for nota in notas:
    print(f"Nota: {nota}")

print()

# 2. Percorrendo usando 'for x in range(len(lista))' para acesso por √≠ndices
print("Percorrendo com √≠ndices:")
for x in range(len(notas)):
    print(f"√çndice {x}: {notas[x]}")

print()

# 3. Fatiamento para obter os 3 primeiros elementos
# lista[0:3] ou lista[:3] - do in√≠cio at√© √≠ndice 3 (exclusivo)
primeiros_tres = notas[0:3]
print(f"3 primeiros: {primeiros_tres}")

# 4. Fatiamento para obter os 3 √∫ltimos elementos
# lista[-3:] - dos √∫ltimos 3 elementos at√© o final
ultimos_tres = notas[-3:]
print(f"3 √∫ltimos: {ultimos_tres}")

# 5. Fatiamento com passo para obter elementos de 2 em 2 (√≠ndices pares)
# lista[0:8:2] - do √≠ndice 0 at√© 8, de 2 em 2
elementos_pares = notas[0:8:2]
print(f"Elementos de 2 em 2: {elementos_pares}")

# 6. Fatiamento reverso para inverter a lista
# lista[-1::-1] - do √∫ltimo elemento at√© o primeiro, passo -1 (invertido)
lista_invertida = notas[-1::-1]
print(f"Lista invertida: {lista_invertida}")
```

**üí° Observa√ß√µes importantes**:

- `for item in lista:` itera diretamente sobre os elementos (mais simples)
- `for x in range(len(lista)):` permite acesso por √≠ndices (√∫til quando precisamos do √≠ndice)
- Fatiamento: `lista[in√≠cio:fim:passo]` onde in√≠cio √© inclusivo, fim √© exclusivo
- `lista[:]` cria uma c√≥pia da lista completa
- Fatiamento reverso usa passo negativo: `[-1::-1]` ou `[::-1]`

**üéØ Varia√ß√µes poss√≠veis**:

- Podemos usar `enumerate()`: `for indice, nota in enumerate(notas):`
- Podemos usar `lista[::-1]` como atalho para inverter
- Fatiamento n√£o modifica a lista original, cria uma nova lista

### Solu√ß√£o 1.4 - Trabalhando com Listas Aninhadas

**Explica√ß√£o**:
Este exerc√≠cio trabalha com estruturas bidimensionais (listas dentro de listas), demonstrando acesso por m√∫ltiplos √≠ndices, modifica√ß√£o de elementos aninhados, itera√ß√£o com la√ßos aninhados e desempacotamento.

**Conceitos aplicados**:
- Cria√ß√£o de listas aninhadas
- Acesso usando m√∫ltiplos √≠ndices: `lista[linha][coluna]`
- Modifica√ß√£o de elementos aninhados
- Itera√ß√£o com la√ßos aninhados
- Desempacotamento de listas
- Uso de fatiamento para pular elementos

```python
# Criando lista aninhada de alunos
alunos = [
    ['Nome', 'Idade', 'Nota'],
    ['Ana', 20, 8.5],
    ['Bruno', 19, 7.0],
    ['Carla', 21, 9.0],
    ['Diego', 20, 6.5]
]

# 1. Acessando e exibindo a idade de Bruno (linha 2, coluna 1)
# Bruno est√° na linha de √≠ndice 2 (ap√≥s cabe√ßalho e Ana)
idade_bruno = alunos[2][1]
print(f"Idade de Bruno: {idade_bruno}")

# 2. Acessando e exibindo a nota de Carla usando √≠ndices negativos
# Carla est√° na linha -2 (pen√∫ltima), nota est√° na coluna -1 (√∫ltima)
nota_carla = alunos[-2][-1]
print(f"Nota de Carla: {nota_carla}")

# 3. Modificando a nota de Diego para 7.5
# Diego est√° na linha 4 (√≠ndice 4), nota est√° na coluna 2
alunos[4][2] = 7.5

# 4. Percorrendo a lista aninhada usando la√ßos aninhados
print("\nTodos os dados:")
for linha in alunos:
    for dado in linha:
        print(dado, end=" ")
    print()  # Nova linha ap√≥s cada linha de dados

# 5. Percorrendo apenas os dados dos alunos (pulando o cabe√ßalho)
# Usando fatiamento [1:] para pular o primeiro elemento (cabe√ßalho)
print("\nDados dos alunos:")
for dados_aluno in alunos[1:]:
    nome, idade, nota = dados_aluno  # Desempacotamento
    print(f"{nome} tem {idade} anos e nota {nota}")

# 6. Adicionando um novo aluno usando append()
novo_aluno = ['Elena', 22, 8.0]
alunos.append(novo_aluno)

print(f"\nLista ap√≥s adicionar Elena: {alunos}")
```

**üí° Observa√ß√µes importantes**:

- Listas aninhadas criam estruturas bidimensionais (como matrizes)
- Acesso usa m√∫ltiplos √≠ndices: `lista[linha][coluna]`
- √çndices negativos funcionam em ambos os n√≠veis
- Fatiamento `[1:]` √© √∫til para pular cabe√ßalhos ou primeiros elementos
- Desempacotamento facilita trabalhar com dados estruturados
- `append()` adiciona uma nova linha completa √† lista aninhada

**üéØ Varia√ß√µes poss√≠veis**:

- Podemos usar `range(len())` para iterar por √≠ndices: `for i in range(len(alunos)):`
- Podemos usar list comprehension para transformar dados: `[aluno[0] for aluno in alunos[1:]]`
- Podemos usar f-strings para formata√ß√£o mais complexa

### Solu√ß√£o 2.1 - Criando e Acessando Tuplas

**Explica√ß√£o**:
Este exerc√≠cio introduz tuplas, estruturas imut√°veis similares a listas, mas que n√£o podem ser modificadas ap√≥s cria√ß√£o.

**Conceitos aplicados**:
- Cria√ß√£o de tuplas com par√™nteses
- Verifica√ß√£o de tipo com `type()`
- Acesso por √≠ndices (positivos e negativos)
- Operador `in` para verifica√ß√£o
- Fun√ß√£o `len()` para tamanho

```python
# Criando tupla de coordenadas
coordenadas = (10, 20, 30)

# 1. Exibindo a tupla completa
print(f"Tupla: {coordenadas}")

# 2. Verificando o tipo da vari√°vel
print(f"Tipo: {type(coordenadas)}")

# 3. Acessando e exibindo o primeiro elemento usando √≠ndice positivo (0)
print(f"Primeiro elemento: {coordenadas[0]}")

# 4. Acessando e exibindo o √∫ltimo elemento usando √≠ndice negativo (-1)
print(f"√öltimo elemento: {coordenadas[-1]}")

# 5. Verificando se o valor 20 est√° na tupla usando 'in'
esta_na_tupla = 20 in coordenadas
print(f"20 est√° na tupla: {esta_na_tupla}")

# 6. Exibindo o tamanho da tupla usando len()
tamanho = len(coordenadas)
print(f"Tamanho: {tamanho}")
```

**üí° Observa√ß√µes importantes**:

- Tuplas s√£o criadas com par√™nteses `()` ou apenas v√≠rgulas
- Tuplas s√£o imut√°veis - n√£o podem ser modificadas ap√≥s cria√ß√£o
- Acesso por √≠ndices funciona igual √†s listas
- Tuplas s√£o mais eficientes em mem√≥ria que listas
- Use tuplas para dados que n√£o devem mudar (coordenadas, configura√ß√µes fixas)

**üéØ Varia√ß√µes poss√≠veis**:

- Podemos criar tupla sem par√™nteses: `coordenadas = 10, 20, 30`
- Tupla com um elemento precisa v√≠rgula: `tupla = (10,)` (n√£o `(10)`)
- Podemos converter lista para tupla: `tuple([1, 2, 3])`

### Solu√ß√£o 2.2 - Modificando Tuplas e Desempacotamento

**Explica√ß√£o**:
Este exerc√≠cio demonstra como modificar tuplas (convertendo para lista e voltando) e o desempacotamento de tuplas em vari√°veis individuais.

**Conceitos aplicados**:
- Imutabilidade de tuplas (tentativa de modifica√ß√£o gera erro)
- Convers√£o tupla ‚Üî lista
- Desempacotamento simples
- Desempacotamento com `*` para capturar m√∫ltiplos elementos

```python
# Tupla original
cores = ('vermelho', 'azul', 'verde', 'amarelo')

# 1. Tentando modificar diretamente (isso gera erro - TypeError)
# Descomente a linha abaixo para ver o erro:
# cores[0] = 'rosa'  # TypeError: 'tuple' object does not support item assignment

# 2. Convertendo tupla para lista, modificando e convertendo de volta
# Passo 1: Converter para lista
cores_lista = list(cores)
# Passo 2: Modificar o segundo elemento (√≠ndice 1)
cores_lista[1] = 'roxo'
# Passo 3: Converter de volta para tupla
cores_modificada = tuple(cores_lista)
print(f"Tupla modificada: {cores_modificada}")

# 3. Desempacotando a tupla original em 4 vari√°veis
cor1, cor2, cor3, cor4 = cores
print(f"Desempacotamento: cor1={cor1}, cor2={cor2}, cor3={cor3}, cor4={cor4}")

# 4. Criando nova tupla com mais elementos
cores_extras = ('laranja', 'rosa', 'preto', 'branco', 'cinza')

# 5. Desempacotando usando * para capturar elementos restantes
# *resto captura todos os elementos restantes em uma lista
c1, c2, *resto = cores_extras
print(f"Desempacotamento com *: c1={c1}, c2={c2}, resto={resto}")
```

**üí° Observa√ß√µes importantes**:

- Tuplas s√£o imut√°veis - tentar modificar gera `TypeError`
- Para modificar: converter para lista ‚Üí modificar ‚Üí converter para tupla
- Desempacotamento permite atribuir elementos a vari√°veis individuais
- `*resto` captura todos os elementos restantes em uma lista
- O n√∫mero de vari√°veis deve corresponder ao n√∫mero de elementos (ou usar `*`)

**üéØ Varia√ß√µes poss√≠veis**:

- Podemos ignorar elementos: `cor1, _, cor3, _ = cores` (usa `_` para ignorar)
- Podemos usar `*` no in√≠cio: `*inicio, ultimo = tupla`
- Podemos usar `*` no meio: `primeiro, *meio, ultimo = tupla`

### Solu√ß√£o 2.3 - Opera√ß√µes com Tuplas

**Explica√ß√£o**:
Este exerc√≠cio demonstra opera√ß√µes com tuplas: concatena√ß√£o, itera√ß√£o, contagem de elementos e busca de √≠ndices.

**Conceitos aplicados**:
- Concatena√ß√£o de tuplas com operador `+`
- Itera√ß√£o sobre tuplas
- M√©todo `count()` para contar ocorr√™ncias
- M√©todo `index()` para encontrar posi√ß√£o
- Fun√ß√£o `len()` para tamanho

```python
# Tuplas iniciais
frutas1 = ('ma√ß√£', 'banana', 'laranja')
frutas2 = ('uva', 'morango', 'abacaxi', 'kiwi')

# 1. Concatenando as duas tuplas usando operador +
# O operador + cria uma nova tupla com todos os elementos
todas_frutas = frutas1 + frutas2
print(f"Tupla concatenada: {todas_frutas}")

# 2. Percorrendo a tupla usando for
print("\nFrutas na tupla:")
for fruta in todas_frutas:
    print(fruta)

# 3. Criando tupla com elementos duplicados
numeros = (1, 2, 3, 2, 4, 2, 5)

# 4. Usando count() para contar quantas vezes o n√∫mero 2 aparece
quantidade_dois = numeros.count(2)
print(f"\nQuantidade de 2: {quantidade_dois}")

# 5. Usando index() para encontrar a posi√ß√£o do primeiro 'abacaxi'
# index() retorna o √≠ndice da primeira ocorr√™ncia
indice_abacaxi = todas_frutas.index('abacaxi')
print(f"√çndice de 'abacaxi': {indice_abacaxi}")

# 6. Verificando o tamanho usando len()
tamanho = len(todas_frutas)
print(f"Tamanho: {tamanho}")
```

**üí° Observa√ß√µes importantes**:

- Concatena√ß√£o com `+` cria uma nova tupla (n√£o modifica as originais)
- Tuplas permitem elementos duplicados
- `count(valor)` retorna o n√∫mero de ocorr√™ncias do valor
- `index(valor)` retorna o √≠ndice da primeira ocorr√™ncia (gera erro se n√£o encontrar)
- Itera√ß√£o sobre tuplas funciona igual √†s listas

**üéØ Varia√ß√µes poss√≠veis**:

- Podemos multiplicar tuplas: `tupla * 3` cria tupla com elementos repetidos
- Podemos usar `index(valor, inicio)` para buscar a partir de um √≠ndice
- Podemos verificar exist√™ncia antes de usar `index()`: `if valor in tupla:`

### Solu√ß√£o 3.1 - Criando e Acessando Dicion√°rios

**Explica√ß√£o**:
Este exerc√≠cio introduz dicion√°rios, estruturas que armazenam pares chave-valor, permitindo acesso r√°pido por chave.

**Conceitos aplicados**:
- Cria√ß√£o de dicion√°rios com chaves `{}`
- Acesso a valores por chave usando colchetes
- Modifica√ß√£o de valores existentes
- Adi√ß√£o de novos pares chave-valor
- Fun√ß√£o `len()` para quantidade de pares

```python
# Criando dicion√°rio pessoa
pessoa = {
    'nome': 'Maria Silva',
    'idade': 28,
    'altura': 1.65,
    'habilitado': True
}

# 1. Exibindo o dicion√°rio completo
print(f"Dicion√°rio: {pessoa}")

# 2. Verificando o tipo
print(f"Tipo: {type(pessoa)}")

# 3. Acessando e exibindo cada valor individualmente usando colchetes
print(f"Nome: {pessoa['nome']}")
print(f"Idade: {pessoa['idade']}")
print(f"Altura: {pessoa['altura']}")
print(f"Habilitado: {pessoa['habilitado']}")

# 4. Modificando a idade para 29
pessoa['idade'] = 29

# 5. Adicionando uma nova chave 'cidade' com valor 'S√£o Paulo'
pessoa['cidade'] = 'S√£o Paulo'

# 6. Exibindo o tamanho do dicion√°rio usando len()
tamanho = len(pessoa)
print(f"\nTamanho: {tamanho}")
print(f"Dicion√°rio atualizado: {pessoa}")
```

**üí° Observa√ß√µes importantes**:

- Dicion√°rios usam chaves `{}` e pares `chave: valor`
- Acesso por chave: `dicionario['chave']` (gera erro se chave n√£o existir)
- Modifica√ß√£o: `dicionario['chave'] = novo_valor`
- Adi√ß√£o: `dicionario['nova_chave'] = valor` (mesma sintaxe)
- `len()` retorna o n√∫mero de pares chave-valor
- Chaves devem ser imut√°veis (strings, n√∫meros, tuplas)

**üéØ Varia√ß√µes poss√≠veis**:

- Podemos usar `get('chave')` para acesso seguro (retorna None se n√£o existir)
- Podemos usar `get('chave', 'padr√£o')` para valor padr√£o
- Podemos criar dicion√°rio vazio: `dicionario = {}` ou `dict()`

### Solu√ß√£o 3.2 - Manipulando Dicion√°rios

**Explica√ß√£o**:
Este exerc√≠cio demonstra m√©todos avan√ßados de manipula√ß√£o de dicion√°rios: acesso seguro com `get()`, verifica√ß√£o de exist√™ncia, atualiza√ß√£o m√∫ltipla e remo√ß√£o de elementos.

**Conceitos aplicados**:
- M√©todo `get()` para acesso seguro
- Operador `in` para verificar exist√™ncia de chave
- M√©todo `update()` para atualizar m√∫ltiplos valores
- M√©todo `pop()` para remover e retornar valor
- M√©todo `popitem()` para remover √∫ltimo item

```python
# Dicion√°rio inicial
produto = {
    'nome': 'Notebook',
    'preco': 2500.00,
    'estoque': 10
}

# 1. Usando get('preco') para acessar o pre√ßo (m√©todo seguro)
# get() retorna None se a chave n√£o existir (n√£o gera erro)
preco = produto.get('preco')
print(f"Pre√ßo (get): {preco}")

# 2. Usando get('desconto') para tentar acessar chave que n√£o existe
# Retorna None em vez de gerar erro
desconto = produto.get('desconto')
print(f"Desconto (get): {desconto}")

# 3. Verificando se 'estoque' existe no dicion√°rio usando 'in'
existe_estoque = 'estoque' in produto
print(f"'estoque' existe: {existe_estoque}")

# 4. Usando update() para atualizar pre√ßo e adicionar categoria
# update() pode atualizar m√∫ltiplos valores de uma vez
produto.update({
    'preco': 2200.00,
    'categoria': 'Eletr√¥nicos'
})

# 5. Usando pop('estoque') para remover e retornar o valor de estoque
# pop() remove a chave e retorna seu valor
estoque_removido = produto.pop('estoque')
print(f"Estoque removido: {estoque_removido}")

# 6. Usando popitem() para remover o √∫ltimo item inserido
# popitem() remove e retorna uma tupla (chave, valor)
ultimo_item = produto.popitem()
print(f"√öltimo item removido: {ultimo_item}")

print(f"\nDicion√°rio final: {produto}")
```

**üí° Observa√ß√µes importantes**:

- `get(chave)` √© mais seguro que `dicionario['chave']` - n√£o gera erro se chave n√£o existir
- `get(chave, padr√£o)` permite especificar valor padr√£o se chave n√£o existir
- `in` verifica exist√™ncia de chave (n√£o de valor)
- `update({chave: valor})` pode atualizar m√∫ltiplos valores simultaneamente
- `pop(chave)` remove e retorna o valor (gera erro se chave n√£o existir)
- `popitem()` remove √∫ltimo item inserido (Python 3.7+ mant√©m ordem de inser√ß√£o)

**üéØ Varia√ß√µes poss√≠veis**:

- Podemos usar `del dicionario['chave']` para remover sem retornar valor
- Podemos usar `clear()` para limpar todos os itens
- Podemos usar `update()` com outro dicion√°rio: `dict1.update(dict2)`

### Solu√ß√£o 3.3 - Percorrendo Dicion√°rios

**Explica√ß√£o**:
Este exerc√≠cio demonstra diferentes formas de iterar sobre dicion√°rios: por chaves, valores e pares chave-valor, al√©m de criar c√≥pias independentes.

**Conceitos aplicados**:
- Itera√ß√£o por chaves (padr√£o e com `keys()`)
- Itera√ß√£o por valores com `values()`
- Itera√ß√£o por pares com `items()`
- M√©todo `copy()` para criar c√≥pia independente
- Diferen√ßa entre refer√™ncia e c√≥pia

```python
# Dicion√°rio de aluno
aluno = {
    'nome': 'Jo√£o',
    'idade': 22,
    'curso': 'Engenharia',
    'nota': 8.5
}

# 1. Percorrendo o dicion√°rio usando 'for chave in dicionario:'
# Por padr√£o, itera sobre as chaves
print("Chaves:")
for chave in aluno:
    print(chave, end=", ")
print()

# 2. Percorrendo e exibindo apenas os valores usando values()
print("\nValores:")
for valor in aluno.values():
    print(valor, end=", ")
print()

# 3. Percorrendo e exibindo chaves e valores usando items()
# items() retorna tuplas (chave, valor)
print("\nChaves e valores:")
for chave, valor in aluno.items():
    print(f"{chave}: {valor}")

# 4. Usando keys() para obter todas as chaves e exibir
chaves = aluno.keys()
print(f"\nTodas as chaves: {chaves}")

# 5. Criando uma c√≥pia do dicion√°rio usando copy()
# copy() cria uma c√≥pia independente (shallow copy)
aluno_copia = aluno.copy()

# Modificando a nota na c√≥pia
aluno_copia['nota'] = 9.0

# 6. Verificando que a modifica√ß√£o na c√≥pia n√£o afetou o original
print(f"\nOriginal: {aluno}")
print(f"C√≥pia: {aluno_copia}")
print(f"Modifica√ß√£o na c√≥pia n√£o afetou o original: {aluno['nota'] == 8.5}")
```

**üí° Observa√ß√µes importantes**:

- `for chave in dicionario:` itera sobre chaves (comportamento padr√£o)
- `values()` retorna todos os valores
- `items()` retorna pares (chave, valor) como tuplas
- `copy()` cria c√≥pia independente (shallow copy)
- Atribui√ß√£o direta (`dict2 = dict1`) cria refer√™ncia (modifica√ß√µes afetam ambos)
- `keys()`, `values()` e `items()` retornam objetos especiais (dict_keys, dict_values, dict_items)

**üéØ Varia√ß√µes poss√≠veis**:

- Podemos converter para lista: `list(dicionario.keys())`
- Podemos usar `dict(dicionario)` para criar c√≥pia (equivalente a `copy()`)
- Podemos iterar apenas valores: `for valor in dicionario.values():`

### Solu√ß√£o 3.4 - Trabalhando com Dicion√°rios Aninhados

**Explica√ß√£o**:
Este exerc√≠cio trabalha com estruturas hier√°rquicas (dicion√°rios dentro de dicion√°rios), demonstrando acesso por m√∫ltiplas chaves, modifica√ß√£o, adi√ß√£o e itera√ß√£o com la√ßos aninhados.

**Conceitos aplicados**:
- Cria√ß√£o de dicion√°rios aninhados
- Acesso usando m√∫ltiplas chaves: `dict['chave1']['chave2']`
- Modifica√ß√£o de valores aninhados
- Adi√ß√£o de novos elementos aninhados
- Itera√ß√£o com la√ßos aninhados usando `items()`
- Remo√ß√£o usando `del`

```python
# Criando dicion√°rio aninhado de clientes
clientes = {
    'cliente1': {
        'nome': 'Ana Paula',
        'idade': 32,
        'email': 'ana@email.com',
        'cidade': 'S√£o Paulo'
    },
    'cliente2': {
        'nome': 'Roberto Lima',
        'idade': 45,
        'email': 'roberto@email.com',
        'cidade': 'Rio de Janeiro'
    },
    'cliente3': {
        'nome': 'Mariana Rocha',
        'idade': 28,
        'email': 'mariana@email.com',
        'cidade': 'Curitiba'
    }
}

# 1. Acessando e exibindo a idade do cliente1 usando m√∫ltiplas chaves
idade_cliente1 = clientes['cliente1']['idade']
print(f"Idade do cliente1: {idade_cliente1}")

# 2. Acessando e exibindo o email do cliente2
email_cliente2 = clientes['cliente2']['email']
print(f"Email do cliente2: {email_cliente2}")

# 3. Modificando a idade do cliente1 para 33
clientes['cliente1']['idade'] = 33

# 4. Adicionando um novo cliente 'cliente4'
# Adicionamos um novo dicion√°rio como valor da chave 'cliente4'
clientes['cliente4'] = {
    'nome': 'Diego Costa',
    'idade': 35,
    'email': 'diego@email.com',
    'cidade': 'Belo Horizonte'
}

# 5. Percorrendo o dicion√°rio aninhado usando la√ßos aninhados
print("\nInforma√ß√µes de todos os clientes:")
for chave_cliente, dados_cliente in clientes.items():
    print(chave_cliente)
    # dados_cliente √© um dicion√°rio, ent√£o iteramos sobre ele
    for chave_dado, valor_dado in dados_cliente.items():
        print(f"  {chave_dado}: {valor_dado}")
    print()  # Linha em branco entre clientes

# 6. Removendo o cliente2 usando del
del clientes['cliente2']

print(f"Clientes ap√≥s remover cliente2: {list(clientes.keys())}")
```

**üí° Observa√ß√µes importantes**:

- Dicion√°rios aninhados criam estruturas hier√°rquicas de dados
- Acesso usa m√∫ltiplas chaves: `dict['chave1']['chave2']`
- Modifica√ß√£o: `dict['chave1']['chave2'] = novo_valor`
- Adi√ß√£o: `dict['nova_chave'] = {dados}` adiciona novo dicion√°rio
- Itera√ß√£o aninhada: primeiro la√ßo itera sobre chaves principais, segundo sobre dados internos
- `del dict['chave']` remove o elemento completamente

**üéØ Varia√ß√µes poss√≠veis**:

- Podemos verificar exist√™ncia: `if 'cliente1' in clientes and 'idade' in clientes['cliente1']:`
- Podemos usar `get()` aninhado: `clientes.get('cliente1', {}).get('idade')`
- Podemos criar estruturas mais profundas (3+ n√≠veis)

### Solu√ß√£o 4.1 - Criando e Manipulando Conjuntos

**Explica√ß√£o**:
Este exerc√≠cio introduz conjuntos (sets), estruturas que n√£o permitem duplicatas e s√£o √∫teis para opera√ß√µes de teoria de conjuntos.

**Conceitos aplicados**:
- Cria√ß√£o de conjuntos com `{}`
- M√©todo `add()` para adicionar elementos
- M√©todo `remove()` para remover (gera erro se n√£o existir)
- M√©todo `discard()` para remover (n√£o gera erro)
- Operador `in` para verifica√ß√£o
- Fun√ß√£o `len()` para tamanho

```python
# Criando conjunto de frutas
frutas = {'banana', 'laranja', 'uva'}

# 1. Exibindo o conjunto completo
print(f"Conjunto inicial: {frutas}")

# 2. Verificando o tipo
print(f"Tipo: {type(frutas)}")

# 3. Adicionando 'melancia' usando add()
frutas.add('melancia')
print(f"Ap√≥s adicionar 'melancia': {frutas}")

# 4. Tentando adicionar 'banana' novamente (n√£o deve duplicar)
frutas.add('banana')
print(f"Ap√≥s tentar adicionar 'banana' novamente: {frutas}")  # N√£o muda

# 5. Verificando se 'laranja' est√° no conjunto usando 'in'
esta_no_conjunto = 'laranja' in frutas
print(f"'laranja' est√° no conjunto: {esta_no_conjunto}")

# 6. Removendo 'uva' usando remove()
frutas.remove('uva')
print(f"Ap√≥s remover 'uva': {frutas}")

# 7. Tentando remover 'abacaxi' usando discard() (n√£o deve gerar erro)
frutas.discard('abacaxi')  # N√£o gera erro se n√£o existir
print(f"Ap√≥s tentar remover 'abacaxi' com discard(): {frutas}")

# 8. Exibindo o tamanho do conjunto usando len()
tamanho = len(frutas)
print(f"Tamanho: {tamanho}")
print(f"Conjunto final: {frutas}")
```

**üí° Observa√ß√µes importantes**:

- Conjuntos s√£o criados com `{}` ou `set()`
- Conjuntos n√£o permitem elementos duplicados (automaticamente removidos)
- `add()` adiciona um √∫nico elemento
- `remove(elemento)` gera `KeyError` se elemento n√£o existir
- `discard(elemento)` n√£o gera erro se elemento n√£o existir
- Conjuntos s√£o n√£o ordenados (ordem pode variar)

**üéØ Varia√ß√µes poss√≠veis**:

- Podemos criar conjunto vazio: `conjunto = set()` (n√£o `{}` que √© dicion√°rio)
- Podemos criar conjunto a partir de lista: `set([1, 2, 3])`
- Podemos usar `update()` para adicionar m√∫ltiplos elementos

### Solu√ß√£o 4.2 - Opera√ß√µes B√°sicas com Conjuntos

**Explica√ß√£o**:
Este exerc√≠cio demonstra opera√ß√µes b√°sicas com conjuntos: adicionar m√∫ltiplos elementos, remover elementos e criar c√≥pias independentes.

**Conceitos aplicados**:
- M√©todo `update()` para adicionar m√∫ltiplos elementos
- M√©todo `remove()` e `discard()`
- M√©todo `copy()` para criar c√≥pia independente
- M√©todo `clear()` para limpar conjunto
- Diferen√ßa entre refer√™ncia e c√≥pia

```python
# Conjuntos iniciais
conjunto1 = {'a', 'b', 'c'}
conjunto2 = {'c', 'd', 'e'}

print(f"Conjunto1 inicial: {conjunto1}")
print(f"Conjunto2 inicial: {conjunto2}")

# 1. Usando update() para adicionar os elementos de conjunto2 em conjunto1
# update() adiciona todos os elementos do iter√°vel (conjunto, lista, tupla, etc.)
conjunto1.update(conjunto2)
print(f"Conjunto1 ap√≥s update com conjunto2: {conjunto1}")

# 2. Criando novo conjunto e adicionando usando update()
conjunto3 = {'f', 'g'}
conjunto1.update(conjunto3)
print(f"Conjunto1 ap√≥s update com conjunto3: {conjunto1}")

# 3. Removendo 'a' usando remove()
conjunto1.remove('a')
print(f"Conjunto1 ap√≥s remover 'a': {conjunto1}")

# 4. Usando discard() para tentar remover 'z' (que n√£o existe)
# discard() n√£o gera erro se elemento n√£o existir
conjunto1.discard('z')
print(f"Conjunto1 ap√≥s tentar remover 'z' com discard(): {conjunto1}")

# 5. Criando uma c√≥pia de conjunto1 usando copy()
# copy() cria uma c√≥pia independente (shallow copy)
conjunto_copia = conjunto1.copy()
print(f"C√≥pia: {conjunto_copia}")

# 6. Limpando o conjunto original usando clear()
# clear() remove todos os elementos, mas mant√©m o conjunto vazio
conjunto1.clear()
print(f"Original ap√≥s clear: {conjunto1}")
print(f"C√≥pia (n√£o afetada): {conjunto_copia}")
```

**üí° Observa√ß√µes importantes**:

- `update(iter√°vel)` adiciona todos os elementos do iter√°vel ao conjunto
- `update()` pode receber conjunto, lista, tupla, string (itera sobre caracteres)
- `remove()` gera `KeyError` se elemento n√£o existir
- `discard()` n√£o gera erro se elemento n√£o existir (mais seguro)
- `copy()` cria c√≥pia independente (modifica√ß√µes n√£o afetam a original)
- `clear()` limpa o conjunto mas mant√©m a estrutura (n√£o deleta)

**üéØ Varia√ß√µes poss√≠veis**:

- Podemos usar `update()` com lista: `conjunto.update([1, 2, 3])`
- Podemos usar `del conjunto` para deletar completamente
- Podemos usar `conjunto1 |= conjunto2` como atalho para `update()`

### Solu√ß√£o 4.3 - Opera√ß√µes de Teoria de Conjuntos

**Explica√ß√£o**:
Este exerc√≠cio demonstra opera√ß√µes fundamentais de teoria de conjuntos: uni√£o, interse√ß√£o, diferen√ßa e diferen√ßa sim√©trica, usando tanto m√©todos quanto operadores.

**Conceitos aplicados**:
- Uni√£o: elementos em A ou B (ou ambos)
- Interse√ß√£o: elementos em A e B simultaneamente
- Diferen√ßa: elementos em A mas n√£o em B
- Diferen√ßa sim√©trica: elementos em A ou B, mas n√£o em ambos
- Uso de m√©todos e operadores equivalentes

```python
# Conjuntos num√©ricos
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}
C = {7, 8, 9, 10}

print(f"Conjunto A: {A}")
print(f"Conjunto B: {B}")
print(f"Conjunto C: {C}\n")

# 1. Uni√£o: A ‚à™ B usando m√©todo union() e operador |
# Uni√£o: elementos que est√£o em A ou em B (ou em ambos)
uniao_metodo = A.union(B)
uniao_operador = A | B
print(f"Uni√£o A ‚à™ B (m√©todo): {uniao_metodo}")
print(f"Uni√£o A ‚à™ B (operador): {uniao_operador}")
print(f"M√©todo e operador s√£o equivalentes: {uniao_metodo == uniao_operador}\n")

# 2. Interse√ß√£o: A ‚à© B usando m√©todo intersection() e operador &
# Interse√ß√£o: elementos que est√£o em A e em B simultaneamente
intersecao_metodo = A.intersection(B)
intersecao_operador = A & B
print(f"Interse√ß√£o A ‚à© B (m√©todo): {intersecao_metodo}")
print(f"Interse√ß√£o A ‚à© B (operador): {intersecao_operador}\n")

# 3. Diferen√ßa: A - B usando m√©todo difference() e operador -
# Diferen√ßa: elementos que est√£o em A mas n√£o em B
diferenca_metodo = A.difference(B)
diferenca_operador = A - B
print(f"Diferen√ßa A - B (m√©todo): {diferenca_metodo}")
print(f"Diferen√ßa A - B (operador): {diferenca_operador}\n")

# 4. Diferen√ßa Sim√©trica: A ‚ñ≥ B usando m√©todo symmetric_difference()
# Diferen√ßa sim√©trica: elementos que est√£o em A ou em B, mas n√£o em ambos
diferenca_simetrica = A.symmetric_difference(B)
print(f"Diferen√ßa Sim√©trica A ‚ñ≥ B: {diferenca_simetrica}\n")

# 5. Calculando a interse√ß√£o entre B e C
intersecao_B_C = B.intersection(C)
print(f"Interse√ß√£o B ‚à© C: {intersecao_B_C}\n")

# 6. Calculando a uni√£o de todos os tr√™s conjuntos (A ‚à™ B ‚à™ C)
# Podemos encadear m√©todos ou usar operadores
uniao_todos_metodo = A.union(B).union(C)
uniao_todos_operador = A | B | C
print(f"Uni√£o A ‚à™ B ‚à™ C (m√©todo): {uniao_todos_metodo}")
print(f"Uni√£o A ‚à™ B ‚à™ C (operador): {uniao_todos_operador}")
```

**üí° Observa√ß√µes importantes**:

- **Uni√£o (‚à™)**: elementos em A ou B (ou ambos) - `union()` ou `|`
- **Interse√ß√£o (‚à©)**: elementos em A e B simultaneamente - `intersection()` ou `&`
- **Diferen√ßa (-)**: elementos em A mas n√£o em B - `difference()` ou `-`
- **Diferen√ßa Sim√©trica (‚ñ≥)**: elementos em A ou B, mas n√£o em ambos - `symmetric_difference()`
- M√©todos e operadores s√£o equivalentes (mesmo resultado)
- Opera√ß√µes n√£o modificam os conjuntos originais (criam novos conjuntos)

**üéØ Varia√ß√µes poss√≠veis**:

- Podemos usar `A.update(B)` para modificar A com uni√£o (modifica A)
- Podemos usar `A.intersection_update(B)` para modificar A com interse√ß√£o
- Podemos verificar subconjuntos: `A.issubset(B)` ou `A <= B`
- Podemos verificar superconjuntos: `A.issuperset(B)` ou `A >= B`
