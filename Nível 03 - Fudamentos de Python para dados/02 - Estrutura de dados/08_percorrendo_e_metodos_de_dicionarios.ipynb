# **Percorrendo e métodos de dicionários**

```python
# Percorrendo um dicionário
pessoas = {
    "nome": "Joaquim dos Santos Reis",
    "idade": 19,
    "altura": 1.79,
    "habilitado": False
}

print(pessoas)
```

```python
# Impressão dos valores do dicionário
for k in pessoas:
  print(pessoas[k])
```

```python
# Impressão chave: valor do dicionário
for k in pessoas:
  print(f"{k} : {pessoas[k]}")
```

```python
# Utilizando método keys()
for chave in pessoas.keys():
  print(chave)
```

```python
# Utilizando método values()
for valor in pessoas.values():
  print(valor)
```

```python
# Utilizando método items()
for chave, valor in pessoas.items():
  print(f"{chave} : {valor}")
```
