# **Métodos de sets**

```python
# Criando um conjunto de números
a = {1, 2, 3}
b = {3, 4, 5}
c = {7, 8, 9}
d = {1, 2, 5, 8, 9}

print(a)
print(b)
print(c)
print(d)
```

```python
# Método copy
aux = a.copy()
print(aux)
```

```python
# Método união (Juntar os elementos dos conjuntos)
a = {1, 2, 3}
b = {3, 4, 5}

# 1ª forma
aux = a.union(b)
print(aux)

# 2ª forma
aux = a | b
print(aux)
```

```python
# Interseção (Apenas os elementos em comum entre os conjuntos)
a = {1, 2, 3}
d = {1, 2, 5, 8, 9}

# 1ª forma
aux = a.intersection(d)
print(aux)

# 2ª forma
aux = a & d
print(aux)
```

```python
# Diferença (apenas o que tem no primeiro conjunto e não tem no segundo)
c = {7, 8, 9}
d = {1, 2, 5, 8, 9}

# 1ª forma
aux = c.difference(d)
print(aux)

# 2ª forma
aux = c - d
print(aux)
```

```python
# Diferença simétrica (O que tem no primeiro e no segundo conjunto, mas não tem em ambos)
a = {1, 2, 3}
b = {3, 4, 5}

aux = a.symetric_difference(b)
print(aux)
```
