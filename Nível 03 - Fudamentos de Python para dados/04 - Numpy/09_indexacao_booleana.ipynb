# **Indexação booleana e filtragem condicional**

A indexação booleana é uma técnica poderosa para filtrar elementos de um array com base em uma ou mais condições lógicas. Você cria um array booleano (True/False) e o usa para selecionar elementos correspondentes.

```python
import numpy as np
```

```python
dados = np.arange(1, 11)
print(dados)
```

```python
# Criando uma condição
condicao = dados > 5
print(f"Array de condição: {condicao}")
```

```python
elementos = dados[condicao]
print(elementos)
```

```python
# Elementos pares e maiores que 3
condicao1 = dados % 2 == 0
condicao2 = dados > 3

elementos = dados[condicao1 & condicao2]
print(elementos)
```

```python
# Filtragem em matriz
matriz = np.array([[10, 20, 30], [40, 50, 60], [70, 80, 90]])
print(matriz)
```

```python
# Filtrar elementos maiores que 50
filtro = matriz > 50
print(filtro)

elementos = matriz[filtro]
print(elementos)
```

Cópia x Visão

```python
array_original = np.arange(1, 6)
print(f"Array Original: {array_original}")

# Criar uma visão (fatiamento)

visao = array_original[1:4]
print(f"Array fatiado - Visão: {visao}")

visao[0] = 99

print(visao)
print(array_original)
```

```python
# Criar uma cópia
array_copia = array_original.copy()
print(f"Array cópia do original: {array_copia}")

array_copia[0] = 99

print(array_copia)
print(array_original)
```
