# **Atributos de Arrays**

```python
import numpy as np
```

```python
# dtype - Indica o tipo de dado dos elementos armazenados no array (ex, int64, float64).
array_int = np.array([5, 17, 32, -50])

print(array_int)

print(array_int.dtype)
```

```python
# Array de float
array_float = np.array([2, 1.5, 3.4, -17, -3.2])
print(array_float)
print(array_float.dtype)
```

```python
# Array de booleano
array_bool = np.array([True, False, True])
print(array_bool)
print(array_bool.dtype)
```

```python
# Array de strings
array_str = np.array(["José", "Maria"])
print(array_str)
print(array_str.dtype)
```

```python
# shape - Retorna uma tupla indicando o número de elementos em cada dimensão
array_1_dimensao = np.array([10, 43, 22, 9])

print(array_1_dimensao)
print(array_1_dimensao.shape)
```

```python
matriz = np.array([[10, 49, 37], [-13, 9, -45]])
print(matriz)
print(matriz.shape)
```

```python
# size - Retorna o número total de elementos no array
print(matriz.size)
print(array_1_dimensao.size)
```

```python
# ndim - Retorna o número de dimensões (eixos) do array. É o mesmo que a dimensão do array, conforme vimos anteriormente
array_1d = np.array([2.3, 4.5, 9])
print(array_1d.ndim)

matriz = np.array([[1, 10], [3, 19]])
print(matriz.ndim)
```

```python
# Conversão de Arrays - Você pode converter o tipo de dado de um array usando o método .astype()
array_float = np.array([1.5, 2.7, 3.2])
print(array_float)
print(array_float.dtype)

# cast
array_cast = array_float.astype(int)
print(array_cast)
print(array_cast.dtype)

array_aux = array_cast.astype(float)
print(array_aux)
print(array_aux.dtype)
```
