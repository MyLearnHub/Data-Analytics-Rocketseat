# **Funções de Agregação**

NumPy fornece várias funções para agregar (resumir) dados em um array.

```python
import numpy as np
```

```python
# sum - Calcula a soma de todos os elementos do array.
array1 = np. array([1, 2, 3, 4, 5])
soma = array1.sum()

print(f"O resultado da soma dos elementos do Array1 é: {soma}")
```

```python
# sum com array de 2 dimensões.
matriz = np.array([[10, 20, 30], [5, 10, 15]])
soma2 = matriz.sum()

print(f"Soma dos elementos da matriz: {soma2}")
```

```python
# Soma por coluna.
soma_col = matriz.sum(axis=0)
print(f"Soma por coluna: {soma_col}")
```

```python
# Soma por linha.
soma_linha = matriz.sum(axis=1)
print(f"Soma por linha: {soma_linha}")
```

```python
# mean - Calcula a média de aritmética dos elementos.
media = array1.mean()
print(f"Média do Array1: {media}")

# Média por coluna
media_coluna = matriz.mean(axis=0)
print(f"Média por coluna: {media_coluna}")

# Média por linha
media_linha = matriz.mean(axis=1)
print(f"Média por linha: {media_linha}")
```

```python
# std - Calcula o desvio padrão dos elementos.
desvio_padrao = array1.std()
print(f"Desvio Padrão: {desvio_padrao}")
```

```python
# min - Encontram o menor valor no array.
array2 = np.array([9, 2, 7, 1, 5])
print(array2)

# Valor mínimo
min = array2.min()

# Valor máximo
max = array2.max()

print(f"Menor valor: {min}")
print(f"Maior valor: {max}")
```

```python
matriz2 = np.array([[10, 2, 8], [4, 15, 6]])
print(f"Maior valor: {matriz2.max()}")
print(f"Menor valor: {matriz2.min()}")

# Mínimo por coluna
min_coluna = matriz2.min(axis=0)
print(f"Valor mínimo por coluna: {min_coluna}")

# Máximo por linha
max_linha = matriz2.max(axis=1)
print(f"Valor máximo por linha: {max_linha}")
```

```python
# argmax - Retorna o índice (ou índices) do maior valor ao longo de um eixo.
array3 = np.array([9, 2, 7, 1, 5])
ind_maior = array3.argmax()

print(f"Índice do maior valor: {ind_maior}")

matriz3 = np.array([[10, 2, 8], [4, 15, 6]])
ind_maior = matriz3.argmax()

print(f"Índice de maior valor na matriz: {ind_maior}")
```

```python
# Índice do maior valor por coluna
ind_maior_col = matriz3.argmax(axis=0)
print(f"Índice do maior valor por coluna: {ind_maior_col}")
```

```python
# Índice de maior valor por linha
ind_maior_linha = matriz3.argmax(axis=1)
print(f"Índice do maior valor por linha: {ind_maior_linha}")
```
