# **Operações com Arrays**

```python
import numpy as np
```

Operações Aritméticas

```python
array1 = np.arange(1, 5)
array2 = np.arange(5, 9)

print(array1)
print(array2)

# Soma
soma = array1 + array2
print(f"Soma: {soma}")

# Subtração
subtracao = array1 - array2
print(f"Subtração: {subtracao}")

# Multiplicação
mult = array1 * array2
print(f"Multiplicação: {mult}")

# Divisão
div = array2 / array1
print(f"Divisão: {div}")

# Potência
pot = array1 ** 4
print(f"Potência: {pot}")
```

Operações com Escalar

```python
array = np.array([10, 20, 30, 40, 50, 60])
numero = 5

soma = array + numero
print(soma)

mult = array * numero
print(mult)
```

Broadcasting

```python
array1 = np.array([[1, 2 ,3], [4, 5, 6]])
print(f"Array1: {array1}")
print(array1.shape)

array2 = np.array([10, 20, 30]) # [10, 20, 30] , [10, 20 30]
print(f"Array2: {array2}")
print(array2.shape)

soma = array1 + array2
print(f"Soma do Array1 com Array2 = {soma}")

array3 = np.array([[100], [200]])
print(f"Array3: {array3}")
print(array3.shape)

soma2 = array1 + array3
print(f"Soma entre Array1 e Array3: {soma2}")
```
