# **Funções Úteis**

```python
import numpy as np
```

```python
# Gerar um número aleatório entre 0 e 1
numero = np.random.rand()
print(f"Número aleatório gerado: {numero}")
```

```python
# Gerar um array de 1 dimensão com 5 números aleatórios
array_1d = np.random.rand(5)
print(f"Array com 5 números aleatórios entre 0 e 1: {array_1d}")
```

```python
# Gerar uma matriz 2x3 de números aleatórios entre 0 e 1
matriz_2x3 = np.random.rand(2, 3)
print(f"Matriz 2x3 de números aleatórios entre 0 e 1: \n {matriz_2x3}")
```

```python
# Gerar números inteiros entre 0 e 9
numero = np.random.randint(10)
print(f"Número aleatório entre 0 e 9 gerado: {numero}")
```

```python
# Gerar um array contendo 3 números inteiros aleatórios entre 10 e 20
array_inteiros = np.random.randint(10, 21, size=3)
print(f"Array de inteiros aleatórios entre 10 e 20: {array_inteiros}")
```

```python
# Gerar uma matriz 2x2 de inteiros aleatórios entre 1 e 100
matriz_2x2 = np.random.randint(1, 101, size=(20, 20))
print(f"Matriz 2x2 de números aleatórios entre 1 e 100: \n{matriz_2x2}")
```

```python
# Escolher elementos aleatoriamente de um array existente
array_frutas = np.array(['maçã', 'banana', 'laranja', 'limão', 'abacaxi', 'abacate'])
print(f"Array de frutas: {array_frutas}")
print(f"Shape: {array_frutas.shape}")

fruta = np.random.choice(array_frutas)
print(f"Fruta aleatória escolhida: {fruta}")

frutas = np.random.choice(array_frutas, size=3)
print(f"Array de 3 elementos aleatórios: {frutas}")
```

```python
array_original = np.array([1, 5, 10, 5, 3, 8, 2, 2, 1, 10, 7])
print(f"Array original: {array_original}")
print(f"Shape: {array_original.shape}")
```

```python
# Gerar um array com elementos únicos do array original
array_unico = np.unique(array_original)
print(f"Array original com elementos únicos: {array_unico}")
```

```python
# np.any() - Verifica se há algum TRUE
array1 = np.array([False, False, True, False])
array2 = np.array([False, False, False])

print(np.any(array1))
print(np.any(array2))
```

```python
# np.all() - Verifica se todos são TRUE
array3 = np.array([True, True, True])
array4 = np.array([True, False, True])

print(np.all(array3))
print(np.all(array4))
```

```python
array_numeros = np.array([9, 2, 7, 1, 5, 3])
print(f"Array de números: {array_numeros}")
```

```python
# np.sort() - Retorna uma cópia ordenada do array
array_ordenado = np.sort(array_numeros)
print(f"Array de números: {array_numeros}")
print(f"Array de números ordenado: {array_ordenado}")
```

```python
array_numeros = np.array([9, 2, 7, 1, 5, 3])
print(f"Array de números: {array_numeros}")
array_numeros.sort()
print(f"Array de números: {array_numeros}")
```

```python
# np.argsort() - Retorna os índices que ordenariam
array_numeros = np.array([9, 2, 7, 1, 5, 3])
array_indices = np.argsort(array_numeros)

print(f"Array de índices que ordenariam o array números: \n{array_indices}")
```

```python
# Ordenação de matrizes (linha ou coluna)
matriz = np.array([[3, 1, 2], [6, 5, 4]])
print(f"Matriz não ordenada: \n{matriz}")
```

```python
# Ordenar a matriz por linha
matriz_linha = np.sort(matriz, axis=1)
print(f"Matriz ordenada por linha: \n{matriz_linha}")
```

```python
# Ordenar a matriz por coluna
matriz_coluna = np.sort(matriz, axis=0)
print(f"Matriz ordenada por coluna: \n{matriz_coluna}")
```
