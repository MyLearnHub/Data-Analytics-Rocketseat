# **Operações Lógicas**

```python
import numpy as np
```

```python
array1 = np.array([10, 20, 30, 40, 50])
array2 = np.array([5, 20, 35, 40, 45])

print(f"Array1: {array1}")
print(f"Array2: {array2}")
```

```python
# Igualdade
igual = array1 == array2
print(f"Array de igualdade: {igual}")
```

```python
# Maior que
maior_que = array1 > array2
print(f"Array maior que {maior_que}")
```

```python
# Maior ou igual
maior_ou_igual = array1 >= array2
print(f"Array maior ou igual a {maior_ou_igual}")
```

```python
# Menor que
menor_que = array1 < array2
print(f"Array menor que {menor_que}")
```

```python
# Menor ou igual
menor_ou_igual = array1 <= array2
print(f"Array menor ou igual a {menor_ou_igual}")
```

```python
# Operadores Lógicos
array1 = np.array((True, False, True))
array2 = np.array((False, False, True))
```

```python
# AND
array_and = array1 & array2
print(f"Operador and: {array_and}")

np.logical_and(array1, array2)
```

```python
# OR
array_or = array1 | array2
print(f"Operador or: {array_or}")

np.logical_or(array1, array2)
```

```python
# NOT
array_not = ~array1
print(f"Operador not: {array_not}")

np.logical_not(array1)
```

```python
# Filtrar elementos
dados = np.array([1, 5, 10, 15, 20, 25])

filtro = dados > 10
print(f"Array filtros: {filtro}")
```

```python
# Usar o filtro para selecionar elementos
elementos_filtrados = dados[filtro]
print(f"Elementos filtrados: {elementos_filtrados}")
```
