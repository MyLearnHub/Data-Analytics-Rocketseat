### Solu√ß√£o 1.2 - Fun√ß√£o com Vari√°veis

**Explica√ß√£o**:
Este exerc√≠cio demonstra como usar vari√°veis dentro de fun√ß√µes, incluindo a entrada de dados do usu√°rio atrav√©s de `input()` e a convers√£o de tipos. A fun√ß√£o realiza um c√°lculo e exibe o resultado.

**Conceitos aplicados**:
- Uso de `input()` para entrada de dados
- Convers√£o de tipos com `int()` ou `float()`
- Vari√°veis locais em fun√ß√µes
- C√°lculos matem√°ticos dentro de fun√ß√µes
- Formata√ß√£o de sa√≠da com f-strings

````python
# Definindo a fun√ß√£o que calcula a √°rea do ret√¢ngulo
def calcular_area_retangulo():
    # Solicitando a largura ao usu√°rio e convertendo para float
    largura = float(input("Digite a largura do ret√¢ngulo: "))
    
    # Solicitando a altura ao usu√°rio e convertendo para float
    altura = float(input("Digite a altura do ret√¢ngulo: "))
    
    # Calculando a √°rea (largura √ó altura)
    area = largura * altura
    
    # Exibindo o resultado formatado
    print(f"A √°rea do ret√¢ngulo √©: {area}")

# Chamando a fun√ß√£o
calcular_area_retangulo()
````

**üí° Observa√ß√µes importantes**:

- Vari√°veis criadas dentro da fun√ß√£o s√£o locais e n√£o podem ser acessadas fora dela

**üéØ Varia√ß√µes poss√≠veis**:

### Solu√ß√£o 1.3 - Sistema de Mensagens

**Explica√ß√£o**:
Este exerc√≠cio integra m√∫ltiplas fun√ß√µes em um sistema interativo com menu. Combina defini√ß√£o de fun√ß√µes, loops, estruturas condicionais e entrada de dados do usu√°rio para criar uma aplica√ß√£o simples mas completa.

**Conceitos aplicados**:
- M√∫ltiplas fun√ß√µes trabalhando juntas
- Loop `while True` para menu interativo
- Estruturas condicionais (`if/elif/else`)
- Uso de `break` para sair de loops
- Entrada de dados do usu√°rio

````python
# Fun√ß√£o que exibe o menu de op√ß√µes
def exibir_menu():
    print("\n=== MENU ===")
    print("1. Mensagem de boas-vindas")
    print("2. Mensagem de despedida")
    print("3. Mensagem personalizada")
    print("4. Sair")
    print("============\n")

# Fun√ß√£o que exibe mensagem de boas-vindas
def mensagem_boas_vindas():
    print("Bem-vindo ao sistema!")

# Fun√ß√£o que exibe mensagem de despedida
def mensagem_despedida():
    print("Obrigado por usar o sistema!")

# Fun√ß√£o que exibe mensagem personalizada
def mensagem_personalizada():
    nome = input("Digite seu nome: ")
    print(f"Ol√°, {nome}! Como posso ajudar?")

# Loop principal do sistema
while True:
    # Exibe o menu
    exibir_menu()
    
    # Solicita a escolha do usu√°rio
    escolha = input("Escolha uma op√ß√£o (1-4): ")
    
    # Verifica a escolha e executa a fun√ß√£o correspondente
    if escolha == "1":
        mensagem_boas_vindas()
    elif escolha == "2":
        mensagem_despedida()
    elif escolha == "3":
        mensagem_personalizada()
    elif escolha == "4":
        print("Saindo do sistema...")
        break  # Sai do loop
    else:
        print("Op√ß√£o inv√°lida! Tente novamente.")
````

**üí° Observa√ß√µes importantes**:

- `while True` cria um loop infinito que s√≥ para com `break`
- Cada fun√ß√£o tem uma responsabilidade espec√≠fica (princ√≠pio de responsabilidade √∫nica)
- O menu √© exibido repetidamente at√© o usu√°rio escolher sair
- √â importante validar a entrada do usu√°rio (op√ß√£o inv√°lida)

**üéØ Varia√ß√µes poss√≠veis**:

- Podemos adicionar mais op√ß√µes ao menu
- Podemos usar um dicion√°rio para mapear op√ß√µes para fun√ß√µes
- Podemos adicionar tratamento de erros com try/except

### Solu√ß√£o 2.2 - Fun√ß√£o com M√∫ltiplos Par√¢metros

**Explica√ß√£o**:
Este exerc√≠cio demonstra como definir fun√ß√µes com m√∫ltiplos par√¢metros. A ordem dos par√¢metros importa quando chamamos a fun√ß√£o sem usar nomes de par√¢metros.

**Conceitos aplicados**:
- M√∫ltiplos par√¢metros em fun√ß√µes
- Ordem dos argumentos
- C√°lculos matem√°ticos com par√¢metros
- Formata√ß√£o de n√∫meros decimais

````python
# Definindo a fun√ß√£o com tr√™s par√¢metros
def calcular_media(nota1, nota2, nota3):
    # Calculando a m√©dia aritm√©tica
    media = (nota1 + nota2 + nota3) / 3
    
    # Exibindo o resultado formatado
    print(f"A m√©dia das notas {nota1}, {nota2} e {nota3} √©: {media:.2f}")

# Testando a fun√ß√£o com as notas fornecidas
calcular_media(8.5, 7.0, 9.5)
````

**üí° Observa√ß√µes importantes**:

- M√∫ltiplos par√¢metros s√£o separados por v√≠rgulas
- A ordem dos argumentos deve corresponder √† ordem dos par√¢metros
- `{media:.2f}` formata o n√∫mero com 2 casas decimais

**üéØ Varia√ß√µes poss√≠veis**:

### Solu√ß√£o 2.3 - Usando *args

**Explica√ß√£o**:
O `*args` permite que uma fun√ß√£o receba um n√∫mero vari√°vel de argumentos posicionais. Todos os argumentos s√£o coletados em uma tupla, permitindo flexibilidade na quantidade de valores passados.

**Conceitos aplicados**:
- Par√¢metro `*args` para argumentos vari√°veis
- Itera√ß√£o sobre tuplas
- Soma de m√∫ltiplos valores

````python
# Definindo a fun√ß√£o com *args para receber quantidade vari√°vel de n√∫meros
def somar_numeros(*args):
    # Inicializando a vari√°vel soma
    soma = 0
    
    # Iterando sobre todos os valores em args (que √© uma tupla)
    for numero in args:
        soma += numero
    
    # Retornando o resultado
    return soma

# Testando a fun√ß√£o com diferentes quantidades de argumentos
resultado1 = somar_numeros(1, 2, 3)
print(f"Soma: {resultado1}")

resultado2 = somar_numeros(10, 20, 30, 40, 50)
print(f"Soma: {resultado2}")

resultado3 = somar_numeros(5)
print(f"Soma: {resultado3}")
````

**üí° Observa√ß√µes importantes**:

- `*args` coleta todos os argumentos posicionais em uma tupla
- O asterisco (`*`) √© obrigat√≥rio antes de `args` (o nome pode ser qualquer um, mas `args` √© conven√ß√£o)
- Podemos iterar sobre `args` como qualquer tupla
- Alternativamente, podemos usar `sum(args)` para somar diretamente

**üéØ Varia√ß√µes poss√≠veis**:

- Solu√ß√£o mais concisa: `return sum(args)`
- Podemos combinar `*args` com par√¢metros normais: `def funcao(nome, *args)`

### Solu√ß√£o 2.4 - Fun√ß√£o com **kwargs

**Explica√ß√£o**:
O `**kwargs` permite que uma fun√ß√£o receba um n√∫mero vari√°vel de argumentos nomeados (keyword arguments). Todos os argumentos s√£o coletados em um dicion√°rio, permitindo flexibilidade na quantidade e tipos de informa√ß√µes passadas.

**Conceitos aplicados**:
- Par√¢metro `**kwargs` para argumentos nomeados vari√°veis
- Itera√ß√£o sobre dicion√°rios com `.items()`
- Formata√ß√£o de sa√≠da estruturada

````python
# Definindo a fun√ß√£o com **kwargs para receber argumentos nomeados vari√°veis
def criar_perfil(**kwargs):
    # Exibindo o t√≠tulo
    print("=== Perfil ===")
    
    # Iterando sobre todos os pares chave-valor no dicion√°rio kwargs
    for chave, valor in kwargs.items():
        print(f"{chave}: {valor}")
    
    # Exibindo linha separadora
    print("---")

# Testando a fun√ß√£o com diferentes combina√ß√µes de argumentos nomeados
criar_perfil(nome="Jo√£o", idade=25, cidade="S√£o Paulo", profissao="Desenvolvedor")
criar_perfil(nome="Maria", idade=30, email="maria@email.com")
````

**üí° Observa√ß√µes importantes**:

- `**kwargs` coleta todos os argumentos nomeados em um dicion√°rio
- Os dois asteriscos (`**`) s√£o obrigat√≥rios antes de `kwargs` (o nome pode variar, mas `kwargs` √© conven√ß√£o)
- `.items()` retorna pares (chave, valor) do dicion√°rio
- Argumentos nomeados permitem passar informa√ß√µes em qualquer ordem

**üéØ Varia√ß√µes poss√≠veis**:

- Podemos acessar valores espec√≠ficos: `kwargs.get('nome', 'N√£o informado')`
- Podemos combinar `*args` e `**kwargs`: `def funcao(*args, **kwargs)`
- Podemos verificar se uma chave existe: `if 'email' in kwargs:`

### Solu√ß√£o 3.1 - Fun√ß√£o que Retorna um Valor

**Conceitos aplicados**:
- Palavra-chave `return`

````python
# Definindo a fun√ß√£o que retorna o dobro de um n√∫mero
def dobro(numero):
    return numero * 2

# Testando a fun√ß√£o e armazenando os resultados
resultado1 = dobro(5)
print(f"O dobro de 5 √©: {resultado1}")

resultado2 = dobro(10)
print(f"O dobro de 10 √©: {resultado2}")

resultado3 = dobro(15)
print(f"O dobro de 15 √©: {resultado3}")
````

### Solu√ß√£o 3.2 - Fun√ß√£o que Retorna uma Lista

**Explica√ß√£o**:
Fun√ß√µes podem retornar estruturas de dados complexas como listas. Isso √© √∫til para gerar cole√ß√µes de valores processados ou calculados.

**Conceitos aplicados**:
- Retorno de listas
- Gera√ß√£o de sequ√™ncias com loops
- Uso de `append()` para construir listas

````python
# Definindo a fun√ß√£o que gera os primeiros n n√∫meros pares
def gerar_numeros_pares(n):
    # Lista para armazenar os n√∫meros pares
    pares = []
    
    # Contador para os n√∫meros pares encontrados
    contador = 0
    numero = 2  # Come√ßamos do primeiro n√∫mero par
    
    # Loop at√© termos n n√∫meros pares
    while contador < n:
        pares.append(numero)
        numero += 2  # Pr√≥ximo n√∫mero par
        contador += 1
    
    # Retornando a lista completa
    return pares

# Testando a fun√ß√£o
lista1 = gerar_numeros_pares(5)
print(f"Primeiros 5 n√∫meros pares: {lista1}")

lista2 = gerar_numeros_pares(10)
print(f"Primeiros 10 n√∫meros pares: {lista2}")
````

**üí° Observa√ß√µes importantes**:

- Podemos construir listas dinamicamente dentro de fun√ß√µes
- `append()` adiciona elementos ao final da lista
- Alternativamente, podemos usar list comprehension: `return [2*i for i in range(1, n+1)]`
- Listas retornadas podem ser manipuladas normalmente ap√≥s o retorno

**üéØ Varia√ß√µes poss√≠veis**:

- Solu√ß√£o com range: `return list(range(2, 2*n+1, 2))`
- Podemos usar list comprehension para uma solu√ß√£o mais concisa

### Solu√ß√£o 3.3 - Fun√ß√£o que Retorna um Dicion√°rio

**Explica√ß√£o**:
Dicion√°rios s√£o estruturas ideais para retornar dados estruturados com m√∫ltiplas informa√ß√µes relacionadas. Esta fun√ß√£o cria e retorna um dicion√°rio completo com informa√ß√µes calculadas.

**Conceitos aplicados**:
- Retorno de dicion√°rios
- Cria√ß√£o de estruturas de dados complexas

- Organiza√ß√£o de dados relacionados

````python
# Definindo a fun√ß√£o que cria um produto com informa√ß√µes calculadas
def criar_produto(nome, preco, quantidade):
    # Calculando o valor total em estoque
    total = preco * quantidade
    
    # Criando e retornando o dicion√°rio com todas as informa√ß√µes
    produto = {
        "nome": nome,
        "preco": preco,
        "quantidade": quantidade,
        "total": total
    }
    
    return produto

# Testando a fun√ß√£o criando dois produtos
produto1 = criar_produto("Notebook", 2500.00, 5)
print(produto1)

produto2 = criar_produto("Mouse", 50.00, 20)
print(produto2)
````

**üí° Observa√ß√µes importantes**:

- Dicion√°rios permitem organizar dados relacionados de forma estruturada
- Podemos calcular valores derivados antes de criar o dicion√°rio
- Dicion√°rios retornados podem ser acessados normalmente: `produto1["nome"]`
- Podemos usar f-strings para formatar a exibi√ß√£o: `print(f"Nome: {produto1['nome']}")`

**üéØ Varia√ß√µes poss√≠veis**:

- Podemos retornar o dicion√°rio diretamente sem vari√°vel intermedi√°ria
- Podemos adicionar valida√ß√µes (ex: pre√ßo > 0)
- Podemos usar classes em vez de dicion√°rios para estruturas mais complexas

### Solu√ß√£o 3.4 - Calculadora com Retorno

**Explica√ß√£o**:
Este exerc√≠cio integra m√∫ltiplas fun√ß√µes que trabalham juntas. Cada fun√ß√£o realiza uma opera√ß√£o espec√≠fica, e uma fun√ß√£o principal (`calcular`) atua como um roteador que escolhe qual fun√ß√£o chamar baseado na opera√ß√£o solicitada.

**Conceitos aplicados**:
- M√∫ltiplas fun√ß√µes trabalhando juntas
- Fun√ß√µes que chamam outras fun√ß√µes
- Tratamento de casos especiais (divis√£o por zero)
- Estruturas condicionais para roteamento

````python
# Fun√ß√£o para somar dois n√∫meros
def somar(a, b):
    return a + b

# Fun√ß√£o para subtrair dois n√∫meros
def subtrair(a, b):
    return a - b

# Fun√ß√£o para multiplicar dois n√∫meros
def multiplicar(a, b):
    return a * b

# Fun√ß√£o para dividir dois n√∫meros (com tratamento de divis√£o por zero)
def dividir(a, b):
    if b == 0:
        return "Erro: divis√£o por zero!"
    else:
        return a / b

# Fun√ß√£o principal que roteia para a opera√ß√£o correta
def calcular(num1, num2, operacao):
    if operacao == "+":
        return somar(num1, num2)
    elif operacao == "-":
        return subtrair(num1, num2)
    elif operacao == "*":
        return multiplicar(num1, num2)
    elif operacao == "/":
        return dividir(num1, num2)
    else:
        return "Opera√ß√£o inv√°lida!"

# Testando todas as opera√ß√µes
resultado1 = calcular(10, 5, "+")
print(f"10 + 5 = {resultado1}")

resultado2 = calcular(10, 5, "-")
print(f"10 - 5 = {resultado2}")

resultado3 = calcular(10, 5, "*")
print(f"10 * 5 = {resultado3}")

resultado4 = calcular(10, 5, "/")
print(f"10 / 5 = {resultado4}")

resultado5 = calcular(10, 0, "/")
print(f"10 / 0 = {resultado5}")
````

**üí° Observa√ß√µes importantes**:

- Cada fun√ß√£o tem uma responsabilidade √∫nica (princ√≠pio de responsabilidade √∫nica)
- A fun√ß√£o `calcular` atua como um dispatcher/roteador
- √â importante tratar casos especiais como divis√£o por zero
- Podemos usar dicion√°rios para mapear opera√ß√µes a fun√ß√µes: `ops = {"+": somar, "-": subtrair}`

**üéØ Varia√ß√µes poss√≠veis**:

- Usar dicion√°rio de opera√ß√µes: `ops[operacao](num1, num2)`
- Adicionar mais opera√ß√µes (pot√™ncia, m√≥dulo, etc.)
- Validar se os n√∫meros s√£o v√°lidos antes de calcular

## üìå T√≥pico 4: Fun√ß√µes Lambda

---

### Solu√ß√£o 4.1 - Lambda B√°sica

**Explica√ß√£o**:
Fun√ß√µes lambda s√£o fun√ß√µes an√¥nimas (sem nome) definidas em uma √∫nica linha. S√£o √∫teis para opera√ß√µes simples e r√°pidas. Devem ser atribu√≠das a uma vari√°vel para serem reutilizadas.

**Conceitos aplicados**:
- Sintaxe de fun√ß√µes lambda
- Atribui√ß√£o de lambda a vari√°veis
- Opera√ß√µes matem√°ticas com lambda

````python
# Criando uma fun√ß√£o lambda que eleva um n√∫mero ao cubo
elevar_ao_cubo = lambda x: x ** 3

# Testando a fun√ß√£o lambda com diferentes valores
print(f"2 ao cubo = {elevar_ao_cubo(2)}")
print(f"3 ao cubo = {elevar_ao_cubo(3)}")
print(f"5 ao cubo = {elevar_ao_cubo(5)}")
````

**üí° Observa√ß√µes importantes**:

- Sintaxe: `lambda parametros: expressao`
- Lambda n√£o precisa de `return` expl√≠cito (a express√£o √© automaticamente retornada)
- Lambda √© √∫til para fun√ß√µes simples de uma linha
- Para fun√ß√µes mais complexas, use `def` normal

**üéØ Varia√ß√µes poss√≠veis**:

- Podemos usar lambda diretamente sem atribuir: `(lambda x: x**3)(2)`
- Lambda √© comumente usado com `map()`, `filter()`, `sorted()`

### Solu√ß√£o 4.2 - Lambda com M√∫ltiplos Par√¢metros

**Explica√ß√£o**:
Lambda pode receber m√∫ltiplos par√¢metros, separados por v√≠rgulas. Podemos usar operadores condicionais (tern√°rios) para l√≥gica mais complexa.

**Conceitos aplicados**:
- Lambda com m√∫ltiplos par√¢metros
- Operadores tern√°rios em lambda
- Express√µes condicionais

````python
# Lambda para calcular pot√™ncia (base elevada ao expoente)
calcular_potencia = lambda base, expoente: base ** expoente

# Lambda para calcular per√≠metro do ret√¢ngulo
calcular_perimetro_retangulo = lambda comprimento, largura: 2 * (comprimento + largura)

# Lambda para verificar qual n√∫mero √© maior (usando operador tern√°rio)
verificar_maior = lambda x, y: x if x > y else y

# Testando as fun√ß√µes lambda
print(f"2 elevado a 8 = {calcular_potencia(2, 8)}")
print(f"Per√≠metro do ret√¢ngulo (5x3) = {calcular_perimetro_retangulo(5, 3)}")
print(f"O maior entre 10 e 7 √©: {verificar_maior(10, 7)}")
````

**üí° Observa√ß√µes importantes**:

- M√∫ltiplos par√¢metros s√£o separados por v√≠rgulas: `lambda x, y: expressao`
- Operador tern√°rio: `valor_se_verdadeiro if condicao else valor_se_falso`
- Lambda pode usar qualquer express√£o Python v√°lida
- Para l√≥gica mais complexa, considere usar fun√ß√£o `def` normal

**üéØ Varia√ß√µes poss√≠veis**:

- Podemos usar `max(x, y)` em vez do operador tern√°rio
- Podemos aninhar operadores tern√°rios (mas cuidado com legibilidade)

### Solu√ß√£o 4.3 - Lambda em Lista

**Explica√ß√£o**:
Podemos armazenar fun√ß√µes lambda em listas (ou outras estruturas) e iterar sobre elas, aplicando cada fun√ß√£o aos mesmos dados. Isso permite criar sistemas flex√≠veis de processamento.

**Conceitos aplicados**:
- Armazenamento de fun√ß√µes em listas
- Itera√ß√£o sobre fun√ß√µes
- Aplica√ß√£o de m√∫ltiplas opera√ß√µes aos mesmos dados
- Tratamento de erros em lambda

````python
# Criando uma lista de fun√ß√µes lambda para opera√ß√µes matem√°ticas
operacoes = [
    lambda x, y: x + y,      # Soma
    lambda x, y: x - y,      # Subtra√ß√£o
    lambda x, y: x * y,       # Multiplica√ß√£o
    lambda x, y: x / y if y != 0 else 0  # Divis√£o (com prote√ß√£o contra divis√£o por zero)
]

# Nomes das opera√ß√µes para exibi√ß√£o
nomes_operacoes = ["+", "-", "*", "/"]

# Fun√ß√£o que aplica todas as opera√ß√µes a dois n√∫meros
def aplicar_operacoes(num1, num2):
    resultados = []
    for i, operacao in enumerate(operacoes):
        resultado = operacao(num1, num2)
        print(f"{num1} {nomes_operacoes[i]} {num2} = {resultado}")
        resultados.append(resultado)
    return resultados

# Testando com os n√∫meros 20 e 4
aplicar_operacoes(20, 4)
````

**üí° Observa√ß√µes importantes**:

- Fun√ß√µes s√£o objetos de primeira classe em Python, podem ser armazenadas em listas
- `enumerate()` retorna √≠ndice e valor ao iterar
- Podemos usar operador tern√°rio em lambda para tratamento de erros
- Esta abordagem permite adicionar novas opera√ß√µes facilmente √† lista

**üéØ Varia√ß√µes poss√≠veis**:

- Usar dicion√°rio em vez de lista: `ops = {"+": lambda x,y: x+y, ...}`
- Usar `zip()` para combinar opera√ß√µes e nomes
- Adicionar valida√ß√£o de entrada antes de aplicar opera√ß√µes

## üìå T√≥pico 5: List Comprehension

---

### Solu√ß√£o 5.1 - List Comprehension B√°sica

**Explica√ß√£o**:
List comprehension √© uma forma concisa e elegante de criar listas. Substitui loops tradicionais por uma √∫nica linha, tornando o c√≥digo mais leg√≠vel e eficiente.

**Conceitos aplicados**:
- Sintaxe b√°sica de list comprehension
- Cria√ß√£o de listas a partir de ranges
- Aplica√ß√£o de express√µes a cada elemento

````python
# Usando list comprehension para criar lista com quadrados dos n√∫meros de 1 a 10
quadrados = [i ** 2 for i in range(1, 11)]

# Exibindo a lista criada
print(quadrados)
````

**üí° Observa√ß√µes importantes**:

- Sintaxe: `[expressao for item in iteravel]`
- `range(1, 11)` gera n√∫meros de 1 a 10 (o √∫ltimo valor n√£o √© inclu√≠do)
- List comprehension √© mais eficiente que loops tradicionais
- Equivalente a: `quadrados = []; for i in range(1, 11): quadrados.append(i**2)`

**üéØ Varia√ß√µes poss√≠veis**:

- Podemos usar qualquer iter√°vel: `[x*2 for x in [1,2,3,4,5]]`
- Podemos aplicar fun√ß√µes: `[len(palavra) for palavra in palavras]`

---

### Solu√ß√£o 5.2 - List Comprehension com Condi√ß√£o

**Explica√ß√£o**:
Podemos adicionar condi√ß√µes (`if`) em list comprehension para filtrar elementos. Apenas elementos que satisfazem a condi√ß√£o s√£o inclu√≠dos na lista resultante.

**Conceitos aplicados**:
- List comprehension com filtro condicional
- Uso de `if` para filtrar elementos
- Operador m√≥dulo para verificar paridade

````python
# Usando list comprehension com condi√ß√£o para filtrar n√∫meros √≠mpares
impares = [i for i in range(1, 21) if i % 2 != 0]

# Exibindo a lista criada
print(impares)
````

**üí° Observa√ß√µes importantes**:

- Sintaxe com condi√ß√£o: `[item for item in iteravel if condicao]`
- `i % 2 != 0` verifica se o n√∫mero √© √≠mpar (resto da divis√£o por 2 n√£o √© zero)
- A condi√ß√£o `if` filtra elementos antes de inclu√≠-los na lista
- Podemos usar m√∫ltiplas condi√ß√µes: `[x for x in range(20) if x % 2 == 0 and x > 10]`

**üéØ Varia√ß√µes poss√≠veis**:

- Alternativa: `[i for i in range(1, 21, 2)]` (range com passo 2 come√ßa em 1)
- Podemos combinar m√∫ltiplas condi√ß√µes com `and` ou `or`

### Solu√ß√£o 5.3 - List Comprehension com Strings

**Explica√ß√£o**:
List comprehension pode processar strings e aplicar m√©todos de string a cada elemento. Podemos trabalhar com listas de palavras, caracteres ou qualquer estrutura de dados.

**Conceitos aplicados**:
- List comprehension com strings
- M√©todos de string (`.upper()`, `.split()`)
- Processamento de texto
- M√∫ltiplas list comprehensions

````python
# Frase inicial
frase = "Python √© uma linguagem de programa√ß√£o"

# 1. Criando lista com palavras em mai√∫sculas
palavras_maiusculas = [palavra.upper() for palavra in frase.split()]

# 2. Criando lista com comprimento de cada palavra
comprimentos = [len(palavra) for palavra in frase.split()]

# Exibindo ambas as listas
print(f"Palavras em mai√∫sculas: {palavras_maiusculas}")
print(f"Comprimento das palavras: {comprimentos}")
````

**üí° Observa√ß√µes importantes**:

- `.split()` divide a string em uma lista de palavras (por padr√£o, separa por espa√ßos)
- `.upper()` converte toda a string para mai√∫sculas
- `len()` retorna o comprimento (n√∫mero de caracteres) da string
- Podemos aplicar qualquer m√©todo de string ou fun√ß√£o na express√£o

**üéØ Varia√ß√µes poss√≠veis**:

- Podemos processar caracteres: `[char.upper() for char in "python"]`
- Podemos filtrar palavras: `[p for p in frase.split() if len(p) > 3]`
- Podemos combinar transforma√ß√µes: `[p.upper() for p in frase.split() if len(p) > 2]`

### Solu√ß√£o 5.4 - List Comprehension com if-else

**Explica√ß√£o**:
Podemos usar operadores tern√°rios (`if-else`) em list comprehension para aplicar diferentes transforma√ß√µes baseadas em condi√ß√µes. Isso permite criar listas com valores diferentes dependendo de cada elemento.

**Conceitos aplicados**:
- List comprehension com express√£o condicional (tern√°rio)
- M√∫ltiplas condi√ß√µes aninhadas
- Prioriza√ß√£o de condi√ß√µes
- L√≥gica condicional em list comprehension

````python
# Criando lista com classifica√ß√£o de n√∫meros de 1 a 15
# Prioridade: m√∫ltiplo de 5 > par/√≠mpar
classificacao = [
    "M√∫ltiplo de 5" if i % 5 == 0 else ("Par" if i % 2 == 0 else "√çmpar")
    for i in range(1, 16)
]

# Exibindo a lista criada
print(classificacao)
````

**üí° Observa√ß√µes importantes**:

- Sintaxe: `[valor_se_verdadeiro if condicao else valor_se_falso for item in iteravel]`
- Podemos aninhar operadores tern√°rios para m√∫ltiplas condi√ß√µes
- A ordem das condi√ß√µes importa - verificamos m√∫ltiplo de 5 primeiro
- Para l√≥gica muito complexa, considere usar fun√ß√£o auxiliar

**üéØ Varia√ß√µes poss√≠veis**:

- Usar fun√ß√£o auxiliar para l√≥gica complexa:
  ```python
  def classificar(n):
      if n % 5 == 0: return "M√∫ltiplo de 5"
      return "Par" if n % 2 == 0 else "√çmpar"
  [classificar(i) for i in range(1, 16)]
  ```
