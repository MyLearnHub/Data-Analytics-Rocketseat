# **Introdução a funções**

**Funções:** Funções são blocos de código que executam uma tarefa específica. Elas são definidas uma vez e podem ser reutilizadas sempre que necessário.

```python
# Sintaxe de uma função
def nome_da_funcao():
  print("")
  # bloco de código
```

```python
# Exemplo de uma função que imprime na tela um texto
def imprime_texto():
  print("Módulo de funções em Python!")
```

```python
# Chamando a função
imprime_texto()
```

```python
# Utilizando variáveis com funções
def soma():
  n1 = int(input("Digite o primeiro número: "))
  n2 = int(input("Digite o segundo número: "))
  soma = n1 + n2
  print(f"{n1} + {n2} = {soma}")
```

```python
# Soma
soma()
```

```python
# Posso chamar a função quantas vezes for necessário
soma()
```
