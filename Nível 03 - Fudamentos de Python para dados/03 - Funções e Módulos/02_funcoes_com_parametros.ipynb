# **Funções com parâmetros**

**Parâmetros:** Você pode enviar valores para as funções através dos parâmetros. Esses valores são enviados dentro dos parênteses das funções. Você pode enviar um ou mais valores, desde que separados por vírgula.

```python
# Sintaxe de uma função com parâmetros
def funcao_com_parametros(nome_parametro):
  print(f"{nome_parametro}")

def funcao2_com_parametros(n1, n2, n3):
  print(f"{n1} - {n2} - {n3}")
```

```python
# Chamar uma função com parâmetro
funcao_com_parametros("Funções em Python", "Outro texto")

funcao2_com_parametros(10, 1.88, -54)
```

```python
# Chamar uma função com argumento (Envio de valores de qualquer tipo)
funcao_com_parametros()
funcao_com_parametros(10)
funcao_com_parametros(True)
funcao_com_parametros([50, 30, 67])
funcao_com_parametros((18, 1.33))
```

```python
# Utilizando o *args - Permite o envio de n valores como parâmetros, caso você não saiba a quantidade
def funcao_com_parametros(*valores):
  print(valores)
  print(type(valores))

funcao_com_parametros(10)
funcao_com_parametros(10, 4)
funcao_com_parametros(10, 12, 4, 7)
```

```python
# Função com parâmetros nomeados com quantidade definida - key -> args
def imprime_dados(idade, nome, altura):
  print(f"A pessoa de nome {nome} possui altura {altura} e {idade} anos de idade")

nome = input("Digite o seu nome: ")
idade = int(input("Digite a sua idade: "))
altura = float(input("Digite a sua altura: "))

# imprime_dados(nome, idade, altura)
imprime_dados(nome = nome, idade = idade, altura = altura)
```

```python
# Funções com parâmetros nomeados sem quantidade definida - **kwargs
def funcao_com_key_wargs(**valores):
  print(valores["nome"])
  print(valores["idade"])
  print(valores["cpf"])
```

```python
def funcao_com_key_wargs(**valores):
  for x in valores.keys():
    print(x)

  for y in valores.values():
    print(y)

  for x, y in valores.items():
    print(f"{x}:{y}")
```

```python
funcao_com_key_wargs(nome = "Osvaldo", idade = 33, cpf = "000.000.000-00")
```

# **Valores padrão para os parâmetros**

```python
# Utilizando um valor default
def imprimir_cidade(cidade = "Nenhuma cidade definida"):
  print(cidade)

imprimir_cidade()
imprimir_cidade("Rio de Janeiro")
imprimir_cidade("Fortaleza")
```

```python
# Utilizando parâmetros de funções com lista
def imprimir_lista_frutas(frutas):
  print(frutas)
  print(type(frutas))

  for x in frutas:
    print(x)

lista_frutas = ["banana", "laranja", "abacaxi", "maracujá"]
imprimir_lista_frutas(lista_frutas)
```

```python
# Utilizando parâmetros de funções com dicionários
def imprimir_clientes(clientes):
  print(clientes)
  print(type(clientes))

  for ind, val in clientes.items():
    print(f"{ind} : {val}")

cliente = {
    "nome": "Osvaldo da Silva",
    "idade": 26,
    "cpf": "000.000.000-00"
}

imprimir_clientes(cliente)
```
