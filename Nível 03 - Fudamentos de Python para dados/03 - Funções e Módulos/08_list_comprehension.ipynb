# **List Comprehension**

List Comprehension é uma forma concisa e elegante de criar listas a partir de iteráveis, como listas, ranges e outros.

Ela substitui loops tradicionais por uma única linha de código.

Quando utilizar a list comprehension?
* Quando a operação é simples e direta.
* Quando quer reduzir a quantidade de linhas.
* Para tornar o código mais legível e elegante.

```python
# Sintaxe de list comprehension
[expressão for item in iterável]
```

```python
# Primeiro exemplo básico: Criando uma lista com os quadrados dos números de 1 a 10

# Utilizar listas
lista_quadrados = []

for i in range(1, 11):
  lista_quadrados.append(i ** 2)

print(lista_quadrados)

# Utilizando list comprehension

quadrados = [i ** 2 for i in range(1, 11)]
print(quadrados)
```

```python
# Utilizando if else - Lista de números pares entre 1 e 20

# Utilizando listas
lista_pares = []

for i in range(1, 21):
  if i % 2 == 0:
    lista_pares.append(i)

print(lista_pares)

# Utilizando list comprehension
lista_pares = [i for i in range(1, 21) if i % 2 == 0]
print(lista_pares)
```

```python
# Utilizando com strings - Lista com todas as palavras de uma frase, convertidas para maiúsculas.
frase = "list comprehension python"

lista_palavras = [palavra.upper() for palavra in frase.split()]
print(lista_palavras)
```

```python
# Utilizando list comprehension com if else - Criando lista que diga "Par", "Ímpar" nos números de 1 a 10
lista_par_impar = ["par" if i % 2 == 0 else "impar" for i in range(1, 11)]

print(lista_par_impar)
```
