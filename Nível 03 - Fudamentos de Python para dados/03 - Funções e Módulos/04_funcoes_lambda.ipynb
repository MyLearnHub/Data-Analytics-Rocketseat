# **Funções Lambda**

**Funções Lambda:** Funções Lambda são funções anônimas em Python - ou seja, funções sem nome, criadas para tarefas rápidas e pontuais.

Elas são definidas em uma única linha, usando a palavra-chave lambda.

```python
# Sintaxe de uma função lambda
lambda parâmetros : expressão
```

```python
# Exemplo - Função lambda calculando o quadrado de um número
valor = lambda numero : numero ** 2

print(valor(10))
```

```python
# Efetuando a soma de dois números
soma = lambda a, b : a + b

print(soma(3, 10))
```

```python
# Efetuando a multiplicação entre dois números
mult = lambda a, b : a * b

print(mult(3, 7))
```
