# **Módulos**

**Módulos:** Módulos em Python são arquivos que contém definições e implementações de funções, classes ou variáveis que podem ser reutilizados em outros programas.

Eles ajudam a organiar o código e a evitar a repetição.

```python
# Importação de um módulo Python - biblioteca matemática
import math
```

```python
# Verificando a raiz quadrada de um número
print(math.sqrt(16))
print(math.sqrt(81))
```

```python
# É possível importat apenas uma função dentro de um módulo com o seguinte código
from math import sqrt

print(sqrt(100))
```

## Alias (apelidos)

Apelidos são muito usados para tornar o código mais conciso e legível, especialmente com bibliotecas grandes.

```python
# É possível utilizar alias (apelidos) na importação, para facilitar a utilização
import pandas as pd
import numpy as np
```

```python
# Alias no exemplo da biblioteca matemática
import math as m

print(m.sqrt(36))
```

## Módulo random

```python
# Importação do módulo random
import random
```

```python
# Utilizando uma função para gerar um inteiro aleatório
print(random.randint(1, 50))
```

```python
# Importando apenas uma função dentro do módulo random
from random import randint
```

```python
# Utilizando a função
print(randint(1, 100))
```
