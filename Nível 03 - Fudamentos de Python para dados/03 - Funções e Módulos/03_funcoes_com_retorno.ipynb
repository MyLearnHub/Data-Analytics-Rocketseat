# **Funções com retorno**

```python
# Função retornando valores inteiros
def potencia(valor):
  return valor ** 2

numero = potencia(10)
print(numero)
```

```python
# Função retornando valores de outros tipos
def retorna_string(texto):
  return texto + " Funções com retorno"

# print(retorna_string("Python - Funções com retorno"))

retorno = retorna_string("Python")
print(retorno)
```

```python
# Função retornando estruturas de dados - lista
def frutas():
  return ["banana", "limão", "abacate", "uva"]

lista_frutas = frutas()

print(lista_frutas)
print(type(lista_frutas))
```

```python
# Função retornando estruturas de dados - dicionário
def dados_clientes(nome, idade, cpf = "Não existe cpf cadastrado"):
  return {"nome": "Osvaldo", "idade": 32, "cpf": "111.111.111-11"}

cliente = dados_clientes()

for ind, val in cliente.items():
  print(f"{ind} : {val}")
```

```python
# Bloqueando a utilização de parâmetros nomeados
def funcao_sem_parametros_nomeados(valor, /):
  print(valor)

funcao_sem_parametros_nomeados(valor = 54)
```

```python
# Obrigando a utilização de parâmetros nomeados
def funcao_com_parametros_nomeados(*, valor):
  print(valor)

funcao_com_parametros_nomeados(valor = 4)
```

```python
# Utilizando parâmetros nomeados e não nomeados
def funcao_com_e_sem_parametros_nomeados(x, y, /, *, z, nome, idade):
  print(x)
  print(y)
  print(z)
  print(nome)
  print(idade)

funcao_com_e_sem_parametros_nomeados(10, "Olá", z = 10, nome = "José", idade = 18)
```
