# **Strings**

**STRINGS:** Uma string é uma variável do tipo texto. Ela possui diversos métodos em Python. O tipo de dados de uma string em Python é o str. Vamos conhecer os principais métodos de string.

```python
# Criando uma string em Python
nome = 'João'
nome2 = "José"
```

```python
texto = "Estou aprendendo 'Python'"
print(texto)
```

```python
# Tamanho de uma string
nome = 'Bismark William'
email = 'bismarkwillian@email.com.br'

print(len(nome))
print(len(email))
```

```python
# Operador IN - Verifica a existência de um caracter e retorna um booleano
print('.com.br' in email)
```

```python
# Operador NOT IN - Verifica a existência de um caracter e retorna um booleano
print('bwn' not in email)
```

```python
# Cadeia de caracteres - Strings são imutáveis
texto = "Python"
texto = "Java"
```

```python
# Método capitalize
print(nome)
nome = 'bismark william'
nome - nome.capitalize()
print(nome)
```

```python
# Deixar todas as letras da string em caixa alta
fruta = 'laranja'
print(fruta)
fruta = fruta.upper()
print(fruta)
```

```python
# Deixar todas as letras da string em caixa baixa
fruta = 'MORaNGo'
print(fruta)
fruta = fruta.lower()
print(fruta)
```
