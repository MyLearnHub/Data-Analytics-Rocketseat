# **Saída de dados em Python**

```python
# Comando para saída de dados em Python
print('Olá mundo!')
```

```python
# Utilizando o print para diversos tipos de dados diferentes
print('Bismark')
print(35)
print(1.88)
print(True)
```

```python
# Imprimindo valores de variáveis
email = 'bismark@email.com'
idade = 32
altura = 1.88
print(email)
print(idade)
print(altura)
```

```python
# Imprimindo valores em um único comando print
print(email, idade, altura)
print('Endereço de email: ', email, ' idade: ', idade, ' altura: ', altura)
```

```python
# Utilizando o parâmetro sep
print('Laranja', 'Uva', 'Banana', sep=',')
```

```python
dia = 10
mes = 4
ano = 2025
print(dia, mes, ano, sep='-')
```

```python
# Utilizando o parâmetro end
print('Laranja')
print('Uva')
print('Banana')
```

```python
print('Laranja', end=',')
print('Uva', end=',')
print('Banana', end='.')
```
