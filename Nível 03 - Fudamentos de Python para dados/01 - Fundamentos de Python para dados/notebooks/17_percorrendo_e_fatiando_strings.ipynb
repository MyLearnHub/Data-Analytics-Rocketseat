# **Percorrendo e fatiando Strings**

```python
# Percorrendo uma string com for
texto = 'Python para dados'
for i in texto:
    print(i)
```

```python
# Percorrendo uma string com while
texto = 'Python para dados'
print(texto)

i = 0
while i < len(texto):
    print(texto[i])
    i += 1
```

```python
# Fatiamento de strings
texto = 'Python para dados'

'''
texto[inicio: fim - 1 : passo]
'''
```

```python
# Fatiando uma string
print(texto[0])
print(texto[-1])

print(texto[2:10:1])

print(texto[::])

print(texto[::-1])

print(texto[-1:-6:-1])
```
