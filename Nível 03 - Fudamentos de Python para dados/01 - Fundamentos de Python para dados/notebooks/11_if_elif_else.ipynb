# **If e Else**

```python
# Sintaxe do comando if-elif-else
'''
if condição: (Se a condição for verdadeira)
    Executa esse bloco de comandos
elif condição2: (Se a condição for verdadeira)
    Executa esse bloco de comandos
elif condição3: (Se a condição for verdadeira)
    Executa esse bloco de comandos
else:
    Executa esse bloco de códigos
'''
```

```python
# Exemplo com if-elif-else
idade = 25

if idade < 10:
    print('Você é uma criança!')
elif idade < 18:
    print('Você é um adolescente!')
elif idade < 65:
    print('Você é um adulto!')
else:
    print('Você é um idoso!')
```

```python
# Exemplo utilizando operadores lógicos
idade = 15
tem_carteira = False

if idade >= 18 and tem_carteira:
    print('Você pode dirigir')
elif idade >= 18 and not tem_carteira:
    print('Você não pode dirigir!')
elif idade < 18 and tem_carteira:
    print('Você não pode dirigir por conta da idade')
else:
    print('Você não pode dirigir pela idade e por não ter carteira!')
```
