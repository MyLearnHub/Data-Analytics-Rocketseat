# **Laço de repetição FOR**

**Laço de Repetição FOR:** O for é uma estrutura de repetição usada para percorrer uma sequência de dados como listas, strings, intevalos númericos, entre outros.

Com ele, você pode acessar cada item da sequência, um por vez, e executar uma ação para cada um deles.

```python
# Sintaxe do comando for
'''
for variável in sequência:
    Executa esse bloco de comandos
'''
```

```python
# Exemplo de for com strings
for x in "Python para Dados":
  print(x)
```

```python
# Exemplo de for com listas
nomes = ['Joana', 'Osvaldo', 'Murilo', 'Leticia']

for i in nomes:
    print(i)
```

```python
# For in range - sintaxe
'''
for i in range(inicio, fim, passo)
1) for i in range(fim) inicio = 0, passo = 1
2) for i in range(inicio, fim) passo = 1
3) for i in range(inicio, fim, passo)
'''
```

```python
for i in range(10):
    print(i)
```

```python
for x in range(2, 21):
    print(x)
```

```python
for j in range(2, 21, 2):
    print(j)
```

```python
for i in range(-1, -10, -1):
    print(i)
```

```python
# For com else
# Imprime uma maensagem ao término do loop

for i in range(1, 10):
    print(i)
else:
    print('Fim da execução do for!')
```
