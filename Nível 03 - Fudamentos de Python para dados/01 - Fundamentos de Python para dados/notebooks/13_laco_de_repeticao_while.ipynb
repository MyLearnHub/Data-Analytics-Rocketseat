# **Laço de repetição WHILE**

**Laço de Repetição WHILE:** O laço de repetição while é uma estrutura de repetição que continua executando um bloco de código enquanto a condição definida for verdadeira.

Assim que a condição se torna falsa, o loop para automáticamente.

```python
# Sintaxe do comando while
'''
while condição: (Enquanto a condição for verdadeira)
    Executar o bloco de comandos
'''
```

```python
# Exemplo de utilização com contador de 1 até 10
i = 1

while i <= 10:
  print('Contagem: ', i)
  i += 1
```

```python
# Loop infinito
i = 1

while i <= 10:
  print('Contagem: ', i)
```

```python
# Validação de strings com while
senha = 'abc'

while senha != 'python':
    print('Senha errada! Tente novamente!')

print('Senha correta! O acesso foi liberado!')
```

```python
senha = input('Digite a senha de acesso ao sistema: ')

while senha != 'python':
    senha = input('Senha inválida! Tente novamente!')

print('Senha correta! O acesso foi liberado!')
```

```python
# Usuário tem que digitar um valor entre 1 e 10
numero = int(input('Digite um número entre 1 e 10'))

while numero < 1 or numero > 10:
    numero = int(input('Número inválido! Digite um número entre 1 e 10: '))

print('Número válido! O número digitado foi: ', numero)
```
