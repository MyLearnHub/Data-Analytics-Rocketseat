# **Operadores aritméticos**

**Operadores Aritméticos:** Os operadores aritméticos são utilizados para realizar cálculos matemáticos entre números (Que podem estar dentro de variáveis)

Em Python utilizados os seguintes operadores:
* Soma
* Subtração
* Multiplicação
* Divisão
* Divisão Inteira
* Potência
* Resto da Divisão

```python
# Operador soma
a = 10
b = 5
soma = a + b

print(soma)
```

```python
# Operador subtração
a = 12
b = 5

sub = a - b
print(sub)
```

```python
# Operador multiplicação
a = 6
b = 4

mult = a * b
print(mult)
```

```python
# Operador divisão
a = 16
b = 3

div = a / b
print(div)
```

```python
# Operador divisão inteira
a = 16
b = 3

div2 = a // b
print(div2)
```

```python
# Operador potência
a = 2
b = 3

pot = a ** b
print(pot)
```

```python
# Operador resto da divisão ou módulo
a = 10
b = 4

resto = a % b
print(resto)
```
