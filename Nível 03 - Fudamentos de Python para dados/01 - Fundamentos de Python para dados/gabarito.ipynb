### Solu√ß√£o 1.3 - Calculadora de IMC

**Explica√ß√£o**:
Este exerc√≠cio combina entrada de dados, convers√£o de tipos, opera√ß√µes aritm√©ticas (incluindo pot√™ncia) e formata√ß√£o de n√∫meros decimais. A f√≥rmula do IMC requer elevar a altura ao quadrado.

**Conceitos aplicados**:
- Entrada de dados com `input()`
- Convers√£o para `float()`
- Operador de pot√™ncia `**`
- Formata√ß√£o de n√∫meros decimais

```python
# Solicitando peso e altura
peso = float(input("Digite seu peso em kg: "))
altura = float(input("Digite sua altura em metros: "))

# Calculando o IMC
# IMC = peso / (altura)¬≤
imc = peso / (altura ** 2)

# Exibindo o resultado formatado com 2 casas decimais
print(f"Seu IMC √©: {imc:.2f}")
```

**üí° Observa√ß√µes importantes**:

- O operador `**` √© usado para pot√™ncia: `altura ** 2` √© equivalente a `altura¬≤`
- Podemos tamb√©m usar `altura * altura` para o mesmo resultado
- A formata√ß√£o `:.2f` exibe o n√∫mero com exatamente 2 casas decimais
- A ordem das opera√ß√µes importa: primeiro calculamos a pot√™ncia, depois a divis√£o

**üéØ Varia√ß√µes poss√≠veis**:

- Podemos calcular a pot√™ncia separadamente: `altura_quadrado = altura ** 2` e depois `imc = peso / altura_quadrado`
- Podemos usar par√™nteses para deixar mais claro: `imc = peso / (altura ** 2)`

### Solu√ß√£o 2.1 - Transforma√ß√µes de Texto

**Explica√ß√£o**:
Este exerc√≠cio demonstra os m√©todos b√°sicos de manipula√ß√£o de strings em Python. √â importante entender que strings s√£o imut√°veis - os m√©todos retornam novas strings, n√£o modificam a original.

**Conceitos aplicados**:
- M√©todo `strip()` para remover espa√ßos
- M√©todo `upper()` para mai√∫sculas
- M√©todo `lower()` para min√∫sculas
- M√©todo `capitalize()` para capitalizar

```python
# String original
texto = "  python para dados  "

# Exibindo a string original
print(f"Original: '{texto}'")

# Removendo espa√ßos nas extremidades
texto_sem_espacos = texto.strip()
print(f"Sem espa√ßos: '{texto_sem_espacos}'")

# Convertendo para mai√∫sculas
texto_maiusculas = texto_sem_espacos.upper()
print(f"Mai√∫sculas: '{texto_maiusculas}'")

# Convertendo para min√∫sculas
texto_minusculas = texto_sem_espacos.lower()
print(f"Min√∫sculas: '{texto_minusculas}'")

# Capitalizando (primeira letra mai√∫scula)
texto_capitalizado = texto_sem_espacos.capitalize()
print(f"Capitalizado: '{texto_capitalizado}'")
```

**üí° Observa√ß√µes importantes**:

- Strings s√£o imut√°veis: m√©todos como `upper()` retornam uma nova string, n√£o modificam a original
- `strip()` remove espa√ßos apenas nas extremidades (in√≠cio e fim), n√£o no meio
- `capitalize()` coloca apenas a primeira letra em mai√∫scula, o resto fica min√∫sculo
- Podemos encadear m√©todos: `texto.strip().upper()`

**üéØ Varia√ß√µes poss√≠veis**:

- Podemos usar `title()` para capitalizar cada palavra: `"python para dados".title()` ‚Üí `"Python Para Dados"`
- Podemos usar `swapcase()` para inverter mai√∫sculas/min√∫sculas

### Solu√ß√£o 2.2 - Validador de Email

**Explica√ß√£o**:
Este exerc√≠cio combina operadores de verifica√ß√£o (`in`, `not in`) com m√©todos de string (`endswith()`) para validar um formato de email. √â um exemplo pr√°tico de valida√ß√£o de dados.

**Conceitos aplicados**:
- Operador `in` para verificar exist√™ncia
- Operador `not in` para verificar inexist√™ncia
- M√©todo `endswith()` para verificar termina√ß√£o
- Estruturas condicionais

```python
# Solicitando o email ao usu√°rio
email = input("Digite seu email: ")

# Validando o email
# Verifica se cont√©m @
tem_arroba = "@" in email

# Verifica se termina com .com ou .com.br
termina_valido = email.endswith(".com") or email.endswith(".com.br")

# Verifica se n√£o cont√©m espa√ßos
sem_espacos = " " not in email

# Se todas as condi√ß√µes forem verdadeiras, o email √© v√°lido
if tem_arroba and termina_valido and sem_espacos:
    print("Email v√°lido!")
else:
    print("Email inv√°lido!")
```

- Podemos simplificar usando uma √∫nica condi√ß√£o: `if "@" in email and (email.endswith(".com") or email.endswith(".com.br")) and " " not in email:`

**üéØ Varia√ß√µes poss√≠veis**:

### Solu√ß√£o 2.3 - Extraindo Informa√ß√µes de uma String

**Explica√ß√£o**:
Este exerc√≠cio demonstra o uso de fatiamento (slicing) de strings para extrair partes espec√≠ficas. √â fundamental entender √≠ndices positivos (do in√≠cio) e negativos (do final).

**Conceitos aplicados**:
- Fatiamento de strings `[inicio:fim]`
- √çndices positivos e negativos
- Concatena√ß√£o de strings

```python
# String de data no formato AAAA-MM-DD
data = "2024-03-15"

# Extraindo o ano (primeiros 4 caracteres: √≠ndice 0 a 3)
ano = data[0:4]  # ou simplesmente data[:4]
print(f"Ano: {ano}")

# Extraindo o m√™s (caracteres 5 a 7: √≠ndice 5 a 6)
mes = data[5:7]  # √≠ndice 5 at√© 6 (n√£o inclui o 7)
print(f"M√™s: {mes}")

# Extraindo o dia usando √≠ndices negativos (√∫ltimos 2 caracteres)
# data[-2:] pega do pen√∫ltimo caractere at√© o final
dia = data[-2:]
print(f"Dia: {dia}")

# Formatando a data como DD/MM/AAAA
data_formatada = f"{dia}/{mes}/{ano}"
print(f"Data formatada: {data_formatada}")
```

**üí° Observa√ß√µes importantes**:

- Fatiamento `[inicio:fim]` inclui o √≠ndice `inicio` mas N√ÉO inclui o √≠ndice `fim`
- `data[:4]` √© equivalente a `data[0:4]` (do in√≠cio at√© o √≠ndice 4, excluindo 4)
- `data[-2:]` pega os √∫ltimos 2 caracteres (do √≠ndice -2 at√© o final)
- √çndices negativos contam de tr√°s para frente: -1 √© o √∫ltimo, -2 √© o pen√∫ltimo, etc.

**üéØ Varia√ß√µes poss√≠veis**:

- Podemos usar `split("-")` para dividir a string: `ano, mes, dia = data.split("-")`
- Podemos acessar caracteres individuais: `ano = data[0] + data[1] + data[2] + data[3]` (n√£o recomendado)

### Solu√ß√£o 2.4 - Relat√≥rio Formatado com F-Strings

**Explica√ß√£o**:
Este exerc√≠cio demonstra o uso avan√ßado de f-strings com formata√ß√£o num√©rica, alinhamento de texto, separadores de milhar e porcentagens. √â essencial para criar relat√≥rios profissionais.

**Conceitos aplicados**:
- F-strings com formata√ß√£o
- Formata√ß√£o num√©rica: `:.2f`, `:03`, `:,`
- Formata√ß√£o de porcentagem: `:.2%`
- Alinhamento de texto: `:^20`, `:<20`, `:>20`

```python
# Dados do relat√≥rio
nome_produto = "Notebook Gamer"
quantidade = 125
preco_unitario = 3299.99
desconto = 0.15  # 15%

# Calculando os totais
total_bruto = quantidade * preco_unitario
total_com_desconto = total_bruto * (1 - desconto)

# Exibindo o relat√≥rio formatado
print("=" * 42)
print("RELAT√ìRIO DE VENDAS".center(42))
print("=" * 42)
print(f"Produto:    {nome_produto:^20}")
print(f"Quantidade: {quantidade:03} unidades")
print(f"Pre√ßo Unit.: R$ {preco_unitario:.2f}")
print(f"Desconto:   {desconto:.2%}")
print(f"Total Bruto: R$ {total_bruto:,.2f}")
print(f"Total Final: R$ {total_com_desconto:,.2f}")
print("=" * 42)
```

**üí° Observa√ß√µes importantes**:

- `{texto:^20}` centraliza o texto em 20 caracteres
- `{numero:03}` preenche com zeros √† esquerda at√© ter 3 d√≠gitos
- `{valor:.2f}` formata float com 2 casas decimais
- `{valor:,.2f}` adiciona separador de milhar e 2 decimais
- `{taxa:.2%}` formata como porcentagem com 2 decimais (multiplica por 100 e adiciona %)
- Podemos calcular dentro das f-strings: `{quantidade * preco_unitario:,.2f}`

**üéØ Varia√ß√µes poss√≠veis**:

- Podemos usar `format()` ao inv√©s de f-strings: `"{:.2f}".format(preco_unitario)`
- Podemos definir a largura dinamicamente: `f"{nome_produto:^{largura}}"`

### Solu√ß√£o 3.1 - Calculadora de Opera√ß√µes Aritm√©ticas

**Explica√ß√£o**:
Este exerc√≠cio demonstra todos os operadores aritm√©ticos b√°sicos do Python. √â importante entender a diferen√ßa entre divis√£o normal (`/`) e divis√£o inteira (`//`), al√©m do operador de m√≥dulo (`%`).

**Conceitos aplicados**:
- Operadores aritm√©ticos: `+`, `-`, `*`, `/`, `//`, `%`, `**`
- F-strings para formata√ß√£o

```python
# Definindo os n√∫meros
a = 15
b = 4

# Realizando todas as opera√ß√µes aritm√©ticas
print(f"{a} + {b} = {a + b}")
print(f"{a} - {b} = {a - b}")
print(f"{a} * {b} = {a * b}")
print(f"{a} / {b} = {a / b}")
print(f"{a} // {b} = {a // b}")
print(f"{a} % {b} = {a % b}")
print(f"{a} ** {b} = {a ** b}")
```

**üí° Observa√ß√µes importantes**:

- `//` retorna a parte inteira da divis√£o (divis√£o inteira)
- `%` retorna o resto da divis√£o (m√≥dulo)
- `**` √© o operador de pot√™ncia: `a ** b` significa a elevado a b
- A ordem de preced√™ncia: par√™nteses > pot√™ncia > multiplica√ß√£o/divis√£o > soma/subtra√ß√£o

**üéØ Varia√ß√µes poss√≠veis**:

### Solu√ß√£o 3.2 - Contador com Operadores de Atribui√ß√£o

**Explica√ß√£o**:
Este exerc√≠cio demonstra os operadores de atribui√ß√£o composta, que s√£o formas concisas de atualizar vari√°veis. Eles combinam uma opera√ß√£o aritm√©tica com atribui√ß√£o.

**Conceitos aplicados**:
- Operadores de atribui√ß√£o: `+=`, `-=`, `*=`, `/=`, `//=`, `%=`, `**=`

```python
# Inicializando o contador
contador = 10
print(f"Valor inicial: {contador}")

# Somando 5 (equivalente a: contador = contador + 5)
contador += 5
print(f"Ap√≥s somar 5: {contador}")

# Subtraindo 3 (equivalente a: contador = contador - 3)
contador -= 3
print(f"Ap√≥s subtrair 3: {contador}")

# Multiplicando por 2 (equivalente a: contador = contador * 2)
contador *= 2
print(f"Ap√≥s multiplicar por 2: {contador}")

# Dividindo por 4 (equivalente a: contador = contador / 4)
contador /= 4
print(f"Ap√≥s dividir por 4: {contador}")

# Divis√£o inteira por 2 (equivalente a: contador = contador // 2)
contador //= 2
print(f"Ap√≥s divis√£o inteira por 2: {contador}")

# Resto da divis√£o por 3 (equivalente a: contador = contador % 3)
contador %= 3
print(f"Ap√≥s resto da divis√£o por 3: {contador}")

# Elevando √† pot√™ncia 2 (equivalente a: contador = contador ** 2)
contador **= 2
print(f"Ap√≥s elevar √† pot√™ncia 2: {contador}")
```

**üí° Observa√ß√µes importantes**:

- `contador += 5` √© equivalente a `contador = contador + 5`, mas mais conciso
- Operadores de atribui√ß√£o modificam a vari√°vel diretamente
- Ap√≥s usar `/=`, a vari√°vel se torna float (mesmo que o resultado seja inteiro)
- `//=` mant√©m o tipo se poss√≠vel, mas ap√≥s `/=` j√° √© float

**üéØ Varia√ß√µes poss√≠veis**:

- Podemos usar operadores de atribui√ß√£o com outros valores: `contador += 10`, `contador *= 3`
- Podemos combinar com outras opera√ß√µes: `contador = (contador + 5) * 2`

### Solu√ß√£o 3.3 - Sistema de Valida√ß√£o com Operadores L√≥gicos

**Explica√ß√£o**:
Este exerc√≠cio combina operadores relacionais e l√≥gicos para criar condi√ß√µes complexas. √â fundamental entender como combinar m√∫ltiplas condi√ß√µes com `and`, `or` e `not`.

**Conceitos aplicados**:
- Operadores relacionais: `>=`
- Operadores l√≥gicos: `and`, `not`
- Convers√£o de string para booleano

```python
# Solicitando informa√ß√µes ao usu√°rio
idade = int(input("Digite sua idade: "))
tem_carteira_input = input("Possui carteira de motorista? (True/False): ")
carteira_suspensa_input = input("Est√° com a carteira suspensa? (True/False): ")

# Convertendo strings para booleanos
tem_carteira = tem_carteira_input == "True"
carteira_suspensa = carteira_suspensa_input == "True"

# Verificando se pode dirigir
# Condi√ß√µes: idade >= 18 E tem carteira E N√ÉO est√° suspensa
pode_dirigir = (idade >= 18) and tem_carteira and (not carteira_suspensa)

# Exibindo o resultado
if pode_dirigir:
    print("Pode dirigir!")
else:
    print("N√£o pode dirigir!")
```

**üí° Observa√ß√µes importantes**:

- `and` retorna `True` apenas se TODAS as condi√ß√µes forem verdadeiras
- `not` inverte o valor booleano: `not True` = `False`, `not False` = `True`
- Par√™nteses ajudam a deixar a l√≥gica mais clara, mas n√£o s√£o sempre necess√°rios
- A preced√™ncia √©: `not` > `and` > `or`
- Podemos simplificar: `pode_dirigir = idade >= 18 and tem_carteira and not carteira_suspensa`

### Solu√ß√£o 3.4 - Calculadora de Desconto Progressivo

**Explica√ß√£o**:
Este exerc√≠cio integra m√∫ltiplos conceitos: operadores relacionais para verificar faixas de valores, operadores l√≥gicos para combinar condi√ß√µes, operadores aritm√©ticos para c√°lculos e formata√ß√£o de sa√≠da. Demonstra a preced√™ncia de operadores.

**Conceitos aplicados**:
- Operadores relacionais: `>`, `<=`, `and`
- Operadores l√≥gicos: `and`
- Operadores aritm√©ticos
- Estruturas condicionais aninhadas
- Formata√ß√£o de sa√≠da

```python
# Solicitando dados ao usu√°rio
valor_compra = float(input("Digite o valor da compra: R$ "))
e_vip_input = input("√â cliente VIP? (True/False): ")
e_vip = e_vip_input == "True"

# Determinando o desconto baseado na faixa de valor
desconto = 0.0  # Inicializando desconto

if valor_compra <= 100:
    desconto = 0.0  # Sem desconto
elif valor_compra <= 500:
    desconto = 0.05  # 5% de desconto
elif valor_compra <= 1000:
    desconto = 0.10  # 10% de desconto
else:
    desconto = 0.15  # 15% de desconto

# Adicionando desconto VIP se aplic√°vel
if e_vip:
    desconto += 0.02  # Adiciona 2% de desconto extra

# Calculando valores
valor_desconto = valor_compra * desconto
valor_final = valor_compra - valor_desconto

# Exibindo o resultado
print(f"Valor original: R$ {valor_compra:.2f}")
print(f"Desconto aplicado: {desconto:.2%}")
print(f"Valor do desconto: R$ {valor_desconto:.2f}")
print(f"Valor final: R$ {valor_final:.2f}")
```

**üí° Observa√ß√µes importantes**:

- Usamos `elif` para verificar m√∫ltiplas condi√ß√µes em sequ√™ncia
- A ordem das condi√ß√µes importa: verificamos de menor para maior
- `desconto += 0.02` adiciona 2% ao desconto existente
- `{desconto:.2%}` formata como porcentagem (multiplica por 100 e adiciona %)
- Podemos usar `and` para verificar faixas: `if valor_compra > 100 and valor_compra <= 500:`

**üéØ Varia√ß√µes poss√≠veis**:

- Podemos calcular o desconto total de uma vez: `desconto_total = desconto_faixa + (0.02 if e_vip else 0)`
- Podemos usar operador tern√°rio: `desconto_vip = 0.02 if e_vip else 0`

- A ordem importa: verificamos de menor para maior

**üéØ Varia√ß√µes poss√≠veis**:

### Solu√ß√£o 4.2 - Gerador de Tabuada

**Explica√ß√£o**:
Este exerc√≠cio demonstra o uso do loop `for` com a fun√ß√£o `range()` para iterar sobre uma sequ√™ncia num√©rica. √â um exemplo cl√°ssico de repeti√ß√£o controlada.

**Conceitos aplicados**:
- Loop `for`
- Fun√ß√£o `range()`
- F-strings para formata√ß√£o

```python
# Solicitando o n√∫mero ao usu√°rio
numero = int(input("Digite um n√∫mero para ver sua tabuada: "))

# Exibindo o cabe√ßalho
print(f"\nTabuada do {numero}:")

# Gerando a tabuada de 1 at√© 10
for i in range(1, 11):  # range(1, 11) gera n√∫meros de 1 a 10
    resultado = numero * i
    print(f"{numero} x {i} = {resultado}")
```

**üí° Observa√ß√µes importantes**:

- `range(1, 11)` gera n√∫meros de 1 at√© 10 (o segundo par√¢metro n√£o √© inclu√≠do)
- `range(11)` geraria n√∫meros de 0 a 10
- `range(1, 11, 2)` geraria n√∫meros √≠mpares de 1 a 9 (com passo 2)
- O loop `for` executa o bloco de c√≥digo para cada valor gerado por `range()`
- A vari√°vel `i` assume cada valor da sequ√™ncia em cada itera√ß√£o

**üéØ Varia√ß√µes poss√≠veis**:

- Podemos calcular diretamente na f-string: `print(f"{numero} x {i} = {numero * i}")`
- Podemos usar `range(10)` e ajustar: `for i in range(10): print(f"{numero} x {i+1} = {numero * (i+1)}")`

### Solu√ß√£o 4.3 - Validador de Senha com WHILE

**Explica√ß√£o**:
Este exerc√≠cio demonstra o uso do loop `while` para repetir uma a√ß√£o at√© que uma condi√ß√£o seja satisfeita. Combina valida√ß√£o de strings com m√©todos de verifica√ß√£o e controle de fluxo.

**Conceitos aplicados**:
- Loop `while`
- M√©todos de string: `isupper()`, `islower()`, `isnumeric()`
- Valida√ß√£o de condi√ß√µes

```python
# Vari√°veis para controlar a valida√ß√£o
senha_valida = False

# Loop que repete at√© a senha ser v√°lida
while not senha_valida:
    # Solicitando a senha
    senha = input("Digite uma senha: ")
    
    # Verificando as condi√ß√µes
    tem_minimo_caracteres = len(senha) >= 8
    tem_maiuscula = False
    tem_minuscula = False
    tem_numero = False
    
    # Percorrendo cada caractere da senha para verificar
    for caractere in senha:
        if caractere.isupper():
            tem_maiuscula = True
        if caractere.islower():
            tem_minuscula = True
        if caractere.isnumeric():
            tem_numero = True
    
    # Verificando se todas as condi√ß√µes foram atendidas
    if tem_minimo_caracteres and tem_maiuscula and tem_minuscula and tem_numero:
        senha_valida = True
        print("Senha v√°lida! Cadastro realizado com sucesso.")
    else:
        print("Senha inv√°lida! Tente novamente.")
```

**üí° Observa√ß√µes importantes**:

- `while not senha_valida:` repete enquanto a senha n√£o for v√°lida
- Usamos flags booleanas (`tem_maiuscula`, `tem_minuscula`, etc.) para rastrear condi√ß√µes
- Percorremos a string com `for` para verificar cada caractere
- `isupper()`, `islower()`, `isnumeric()` retornam `True` se o caractere atender ao crit√©rio
- O loop continua at√© que `senha_valida` seja `True`

**üéØ Varia√ß√µes poss√≠veis**:

- Podemos usar `any()` com list comprehension: `tem_maiuscula = any(c.isupper() for c in senha)`
- Podemos verificar todas as condi√ß√µes em uma √∫nica express√£o
- Podemos usar `break` para sair do loop quando a senha for v√°lida

### Solu√ß√£o 4.4 - Menu Interativo com While True

**Explica√ß√£o**:
Este exerc√≠cio integra m√∫ltiplos conceitos: loop infinito com `while True`, controle de fluxo com `break` e `continue`, estruturas condicionais aninhadas e entrada de dados. √â um exemplo completo de programa interativo.

**Conceitos aplicados**:
- `while True` para loop infinito
- `break` para sair do loop
- `continue` para pular itera√ß√£o
- Estruturas condicionais aninhadas
- Entrada e sa√≠da de dados

```python
# Loop infinito controlado
while True:
    # Exibindo o menu
    print("\n=== MENU ===")
    print("1. Calcular √°rea do ret√¢ngulo")
    print("2. Calcular √°rea do c√≠rculo")
    print("3. Verificar se n√∫mero √© par/√≠mpar")
    print("4. Sair")
    
    # Solicitando a op√ß√£o
    opcao = input("Escolha uma op√ß√£o: ")
    
    # Processando a op√ß√£o escolhida
    if opcao == "1":
        # √Årea do ret√¢ngulo: largura √ó altura
        largura = float(input("Digite a largura: "))
        altura = float(input("Digite a altura: "))
        area = largura * altura
        print(f"√Årea do ret√¢ngulo: {area}")
    
    elif opcao == "2":
        # √Årea do c√≠rculo: œÄ √ó raio¬≤
        raio = float(input("Digite o raio: "))
        pi = 3.14159
        area = pi * (raio ** 2)
        print(f"√Årea do c√≠rculo: {area:.2f}")
    
    elif opcao == "3":
        # Verificando se √© par ou √≠mpar
        numero = int(input("Digite um n√∫mero: "))
        if numero % 2 == 0:  # Se o resto da divis√£o por 2 for 0, √© par
            print(f"{numero} √© par")
        else:
            print(f"{numero} √© √≠mpar")
    
    elif opcao == "4":
        # Saindo do programa
        print("Programa encerrado. At√© logo!")
        break  # Interrompe o loop
    
    else:
        # Op√ß√£o inv√°lida
        print("Op√ß√£o inv√°lida! Tente novamente.")
        continue  # Volta ao in√≠cio do loop
```

**üí° Observa√ß√µes importantes**:

- `while True:` cria um loop infinito que s√≥ para com `break`
- `break` interrompe o loop imediatamente e sai dele
- `continue` pula o resto da itera√ß√£o atual e volta ao in√≠cio do loop
- Usamos `elif` para verificar m√∫ltiplas op√ß√µes
- O `else` trata op√ß√µes inv√°lidas
- O operador `%` (m√≥dulo) √© √∫til para verificar paridade: `numero % 2 == 0` significa par

**üéØ Varia√ß√µes poss√≠veis**:

- Podemos usar uma vari√°vel flag: `rodando = True` e `while rodando:` com `rodando = False` para sair
- Podemos adicionar mais op√ß√µes ao menu
- Podemos validar a entrada antes de processar
