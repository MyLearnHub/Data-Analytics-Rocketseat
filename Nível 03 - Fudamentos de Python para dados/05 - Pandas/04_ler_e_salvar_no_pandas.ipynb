# **Ler e salvar no Pandas**

```python
import pandas as pd
```

```python
# Leitura de arquivos csv
url = 'https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv'

df = pd.read_csv(url)
print(df)
```

```python
# Ler arquivos xlsx
df2 = pd.read_excel('exemplo.xlsx')
print(df2)
```

```python
# Salvar no formato csv
df3 = pd.DataFrame({
    'Fruta': ['Maçã', 'Banana', 'Laranja'],
    'Quantidade': [100, 150, 120]
})

df3.to_csv('frutas.csv', index=False)
```

```python
# Lendo o arquivo csv criado
df4 = pd.read_csv('frutas.csv')
print(df4)
```

```python
# Salvando arquivos xlsx
df5 = pd.DataFrame({
    'País': ['Brasil', 'Argentina', 'Chile'],
    'População': [215, 45, 19]
})

print(df5)
df5.to_excel('paises.xlsx', index=False)
```

```python
# Lendo o arquivo xlsx criado
df6 = pd.read_excel('paises.xlsx')
print(df6)
```
