# **astype, isna, fillna e dropna**

```python
import pandas as pd
import numpy as np
```

```python
# Criar o DF
dados = {
    'Numero_String': ['1', '2', '3', '4'],
    'Preco_Float': [10.5, 20.0, 15.75, 30.0]
}

df = pd.DataFrame(dados)
print(df)
```

```python
df.info()
```

```python
# Converter 'Numero_String' para int
df['Numero_String'] = df['Numero_String'].astype(int)
print(df.info())
print(df)
```

```python
# Converter Preco_Float para int
df['Preco_Float'] = df['Preco_Float'].astype(int)
print(df.info())
print(df)
```

```python
# Valores ausentes
df2 = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [np.nan, 2, 3, 4, np.nan],
    'C': [1, 2, 3, 4, 5],
})

print(df2)
```

```python
# isna()
print(df2.isna())
```

```python
# fillna() - Preencher com 0 onde tem valores nulos
df_nulos = df2.fillna({'A': 0})
print(df_nulos)
```

```python
# fillna - Preencher com uma m√©dia de valores
df_media = df2.fillna(df2.mean())
print(df_media)
```

```python
# Remover as linhas com valores ausentes
df_drop_linhas = df2.dropna()
print(df_drop_linhas)
```

```python
df_drop_cols = df2.dropna(axis=1)
print(df_drop_cols)
```
