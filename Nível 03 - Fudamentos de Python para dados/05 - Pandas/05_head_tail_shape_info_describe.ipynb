# **Head, Tail, Shape, Info e Describe**

```python
import pandas as pd
```

```python
# Criando um DF
dados = {
    'Nome': ['Ana', 'Bruno', 'Carla', 'Daniel', 'Eduarda', 'Fernando', 'Gabriela', 'Hugo', 'Isabela', 'Julio'],
    'Idade': [22, 35, 28, 41, 29, 33, 25, 45, 30, 38],
    'Cidade': ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Curitiba', 'Porto Alegre', 'São Paulo', 'Rio de Janeiro', 'Fortaleza', 'Belém', 'Goiania'],
    'Salario': [3000, 5500, 4000, 7000, 4200, 6000, 3500, 8000, 4800, 6500]
}

print(dados)
```

```python
df = pd.DataFrame(dados)
print(df)
```

```python
# head(n) - Retorna as primeiras n linhas do DF
print(df.head(3))
```

```python
# tail() - Retorna as últimas n linhas do DF
print(df.tail(7))
```

```python
# shape - Retorna a quantidade de linhas e colunas
df.shape
```

```python
# info() - Fornece um resumo do DF
df.info()
```

```python
# describe() - Retorna estatísticas sobre o DF
print(df.describe())
```
