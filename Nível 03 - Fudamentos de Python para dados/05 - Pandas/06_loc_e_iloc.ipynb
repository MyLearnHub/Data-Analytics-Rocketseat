# **Loc e Iloc**

```python
import pandas as pd
```

```python
# Criar o DF
dados = {
    'Produto': ['A', 'B', 'C', 'D', 'E'],
    'Preco': [10.5, 20.0, 15.75, 30.0, 25.25],
    'Estoque': [100, 50, 75, 20, 90]
}

index = ['item1', 'item2', 'item3', 'item4', 'item5']

df = pd.DataFrame(dados, index)
print(df)
```

```python
# Visualizar colunas específicas
df['Estoque']
```

```python
# Selecionar múltiplas colunas
print(df[['Produto', 'Estoque']])
```

```python
# Comando loc - Seleção do item3
print(df.loc['item3'])
```

```python
# Comando loc para linhas e colunas
print(df.loc[['item1', 'item5'], ['Preço', 'Estoque']])
```

```python
# Comando iloc para buscar o item3
print(df.iloc[2])
```

```python
# Comando iloc para linhas e colunas
print(df.iloc[0:2, 1:3])
```
