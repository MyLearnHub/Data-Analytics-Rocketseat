# **Formatando colunas**

```python
import pandas as pd
```

```python
dados = {
    'Produto': ['A', 'B', 'C', 'D'],
    'Preco_Unitario': [10, 20, 15, 25],
    'Quantidade_Vendida': [5, 3, 7, 2]
}

df = pd.DataFrame(dados)
print(df)
```

```python
df['Valor_total'] = df['Preco_Unitario'] * df['Quantidade_Vendida']
print(df)
```

```python
# Coluna com valor fixo
df['Desconto_Aplicado'] = 0.10
print(df)
```

```python
# Renomear a coluna produto
df_novo = df.rename(columns={'Produto': 'Produto_Loja'})
print(df_novo)
```

```python
print(df)
```

```python
df.rename(columns={'Produto': 'Produto_Loja', 'Quantidade_Vendida': 'Quantidade_Vendas'}, inplace=True)
print(df)
```

```python
# Removendo colunas
df = df.drop('Desconto_Aplicado', axis=1)
print(df)
```

```python
df.drop(['Quantidade_Vendas', 'Valor_total'], axis=1, inplace=True)
print(df)
```
