# **Séries**

```python
import pandas as pd
```

```python
# Criando uma series a partir de uma lista
numeros = [10, 20, 30, 40, 50]

serie1 = pd.Series(numeros)
print(serie1)
print(type(serie1))
```

```python
# Acessar um elemento da series
print(serie1[3])
```

```python
# Criar uma series com índice personalizado
lista_numeros = [100, 200, 300]
index = ['a', 'b', 'c']

serie_personalizada = pd.Series(lista_numeros, index)
print(serie_personalizada)
print(serie_personalizada['a'])
```

```python
# Criar uma series a partir de um dicionário
dados = {'primeiro': 1, 'segundo': 2, 'terceiro': 3}
print(dados)

serie_dict = pd.Series(dados)
print(serie_dict)

print(serie_dict['terceiro'])
```
